# Levi-Civita Symbol in 3D 

## 1. 核心定義：指標與置換

簡單來說在 $n$ 維空間中 **交換任何兩個指標時都會變號** ，且分量值為 $\pm 1$ 或 $0$ 的張量。 

Levi-Civita symbol 是一個用來描述指標置換關係與空間方向性的符號，Levi-Civita symbol 的本質類似於「完全反對稱張量 (Totally Antisymmetric Tensor)」描述了 **空間中的方向與體積。** 而在三維笛卡兒座標系中其定義如下：

$$\varepsilon_{ijk} = 
\begin{cases} 
+1 & \text{若 } (i,j,k) \text{ 為 } (1,2,3) \text{ 的偶置換（與右手座標系同向）} \\
-1 & \text{若 } (i,j,k) \text{ 為 } (1,2,3) \text{ 的奇置換} \\
0 & \text{若任何兩個指標重複} 
\end{cases}$$

或者將3維的情況列出來：

|$\varepsilon_{ijk}\overset{\text{let}}{=}1$|||
|---|---|---|
|$\varepsilon_{iii}=0$|$\varepsilon_{iij}=0$|$\varepsilon_{iik}=0$|
|$\varepsilon_{iji}=0$|$\varepsilon_{ijj}=0$|$\varepsilon_{ijk}=1$|
|$\varepsilon_{iki}=0$|$\varepsilon_{ikj}=-1$|$\varepsilon_{ikk}=0$|
||||
|$\varepsilon_{jii}=0$|$\varepsilon_{jij}=0$|$\varepsilon_{jik}=-1$|
|$\varepsilon_{jji}=0$|$\varepsilon_{jjj}=0$|$\varepsilon_{jjk}=0$|
|$\varepsilon_{jki}=1$|$\varepsilon_{jkj}=0$|$\varepsilon_{jkk}=0$|
||||
|$\varepsilon_{kii}=0$|$\varepsilon_{kij}=1$|$\varepsilon_{kik}=0$|
|$\varepsilon_{kji}=-1$|$\varepsilon_{kjj}=0$|$\varepsilon_{kjk}=0$|
|$\varepsilon_{kki}=0$|$\varepsilon_{kkj}=0$|$\varepsilon_{kkk}=0$|

>嚴格來說，$\varepsilon_{ijk}$ 在這裡是一個「符號 (Symbol)」。在曲線座標或廣義相對論中真正的 Levi-Civita Tensor 需要額外考慮度量張量的行列式 ($\sqrt{g}$)，但在標準流體力學運算中我們通常在笛卡兒座標下將其視為張量使用。


## 2. 數學本質：行列式的代數化

Levi-Civita symbol將幾何上的有向體積轉化為可運算的指標形式，所以涉及 **旋轉** 與 **垂直** 的物理量像是涉及外積的推導中容易看到Levi-Civita symbol， **能夠讓方程式更簡單地去推導證明** ：

$$\det(\mathbf{A}) = \varepsilon_{ijk} A_{1i} A_{2j} A_{3k}$$

$$(\mathbf{A} \times \mathbf{B})_i = \varepsilon_{ijk} A_j B_k$$


## 假設與已知 (Assumptions & Preliminaries)

* **【假設 1】線性疊加假設** ： 假設複雜的氣候場可以表示為幾個空間型態的 **線性組合** 
  
  * 只關心二階統計，不看偏態、峰態、非線性結構
  * **限制** ：對於強烈非線性、或存在「氣候模態切換 (Regime-switching)」的系統，線性的 EOF 可能會 **失真** ，這時可能需要考慮 Nonlinear PCA 或 Autoencoder。

* **【已知 1】 $\varepsilon-\delta$ 恆等式** ：把 **旋轉（外積）** 變成 **內積（投影）** 的翻譯器
  
  $$\varepsilon_{ijk} \varepsilon_{imn} = \delta_{jm} \delta_{kn} - \delta_{jn} \delta_{km}$$
  
  * 其中 $\delta_{\text{index1 index2}} = \begin{cases} 1, & \text{如果 } \text{index1} = \text{index2} \\0, & \text{如果 } \text{index1} \neq \text{index2} \end{cases}$
