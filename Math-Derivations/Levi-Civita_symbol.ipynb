  **Levi-Civita 符號的本質是「完全反對稱張量 (Totally Antisymmetric Tensor)」，它描述了空間中的方向與體積。**

  簡單來說在 $n$ 維空間中交換任何兩個指標時都會變號、且分量值為 $\pm 1$ 或 $0$ 的張量。
