# 常態分佈高階期望值 $E[x^n]$ 之證明

## 證明目標

對於服從 **常態分佈(Normal distribution or Gaussian distribution)** $x \sim \mathcal{N}(\mu, \sigma^2)$ 的變數 $x$，其 $n$ 階期望值為：

$E[x^n] =\displaystyle \sum_{k=0}^{\lfloor n/2 \rfloor} \binom{n}{2k} \mu^{n-2k} \sigma^{2k} (2k-1)!!$


**常態分佈 $\mathcal{N}(\mu, \sigma^2)$ 之 $n$ 階期望值表**

|$n$|$E[x^n]$|若 μ=0 (僅含 σ 項)|
|---|---|---|
|$1$|$\mu$|$0$|
|$2$|$\mu^2 + \sigma^2$|$\sigma^2$|
|$3$|$\mu^3 + 3\mu\sigma^2$|$0$|
|$4$|$\mu^4 + 6\mu^2\sigma^2 + 3\sigma^4$|$3\sigma^4$|
|$5$|$\mu^5 + 10\mu^3\sigma^2 + 15\mu\sigma^4$|$0$|
|$6$|$\mu^6 + 15\mu^4\sigma^2 + 45\mu^2\sigma^4 + 15\sigma^6$|$15\sigma^6$|
|$7$|$\mu^7 + 21\mu^5\sigma^2 + 105\mu^3\sigma^4 + 105\mu\sigma^6$|$0$|
|$8$|$\mu^8 + 28\mu^6\sigma^2 + 210\mu^4\sigma^4 + 420\mu^2\sigma^6 + 105\sigma^8$|$105\sigma^8$|
|$9$|$\mu^9 + 36\mu^7\sigma^2 + 378\mu^5\sigma^4 + 1260\mu^3\sigma^6 + 945\mu\sigma^8$|$0$|
|$10$|$\mu^{10} + 45\mu^8\sigma^2 + 630\mu^6\sigma^4 + 3150\mu^4\sigma^6 + 4725\mu^2\sigma^8 + 945\sigma^{10}$|$945\sigma^{10}$|
|$n$|$\displaystyle \sum_{k=0}^{\lfloor n/2 \rfloor} \binom{n}{2k} \mu^{n-2k} \sigma^{2k} (2k-1)!!$|$\begin{cases} 0 & n \text{ 為奇數} \\ \sigma^n (n-1)!! & n \text{ 為偶數} \end{cases}$|



## 假設與已知 (Assumptions & Preliminaries)

* **【已知 1】標準常態分佈變數變換**  任何常態分佈 $x \sim \mathcal{N}(\mu, \sigma^2)$ 都可以寫成標準常態分佈 $z \sim \mathcal{N}(0, 1)$ 的線性變換：
  
  $$x = \mu + \sigma z$$
  
  其中 $z$ 的機率密度函數 (PDF) 為 $f(z) = \frac{1}{\sqrt{2\pi}} e^{-\frac{z^2}{2}}$。

* **【已知 2】期望值的線性性質** 對於常數 $a, b$ 與隨機變數 $X, Y$：
  
  $$E[aX + bY] = aE[X] + bE[Y]$$

* **【已知 3】二項式定理 (Binomial Theorem)** 
  
  $$(a+b)^n = \sum_{k=0}^{n} \binom{n}{k} a^{n-k} b^k$$
  
  舉例來說當 $n=6$ 時，巴斯卡三角形係數為：`1, 6, 15, 20, 15, 6, 1`

* **【已知 4】積分邊界條件收斂** 在Integration by Parts運算中，我們知道指數衰減函數 $e^{-\frac{x^2}{2}}$ 在無窮遠處的衰減速度快於任何多項式增長：
  
  $$\lim_{x \to \pm\infty} x^k e^{-\frac{x^2}{2}} = 0 \quad (\text{對於任意正整數 } k)$$
  

## 第一部分：推導標準常態分佈 $z$ 的Moments (動差,矩量)

先處理標準常態分佈 $z$ 的性質，這將是後續計算的基石。

1. **奇數階期望值** 由於標準常態分佈的 PDF $f(z)$ 是對稱於 y 軸，而 $z^n$ 在對稱區間積分為 $0$ 故可得一重要性質：
   
   $$\begin{aligned}
   E[z^n] =& \int_{-\infty}^{\infty} z^n \frac{1}{\sqrt{2\pi}} e^{-\frac{z^2}{2}} dz \\
   =& \int_{-\infty}^{0} z^n \frac{1}{\sqrt{2\pi}} e^{-\frac{z^2}{2}} dz + \int_{0}^{\infty} z^n \frac{1}{\sqrt{2\pi}} e^{-\frac{z^2}{2}} dz \\
   =& 0
   \end{aligned}$$
   
   對於所有奇數 $n$， $E[z^n] = 0 $

2. **偶數階期望值** 的遞迴公式 $E[z^n]$ (當 $n$ 為偶數)。根據期望值定義使用Integration by Parts($z^{n-1} ,z e^{-\frac{z^2}{2}} $)：
   
   $$\begin{aligned}
   E[z^n] =& \int_{-\infty}^{\infty} z^n \frac{1}{\sqrt{2\pi}} e^{-\frac{z^2}{2}} dz \\
   =& \frac{1}{\sqrt{2\pi}} ( \underbrace{\left[ -z^{n-1} e^{-\frac{z^2}{2}} \right]_{-\infty}^{\infty}}_{\text{=0 , 已知4}} - \int_{-\infty}^{\infty} -e^{-\frac{z^2}{2}} (n-1)z^{n-2} dz ) \\
   =& (n-1) \int_{-\infty}^{\infty} z^{n-2} \frac{1}{\sqrt{2\pi}} e^{-\frac{z^2}{2}} dz \\
   =& (n-1) E[z^{n-2}]
   \end{aligned}$$
   
   
   
3. 計算所需 $z$ 的Moments
   
   利用上述遞迴公式與 $E[z^0]=1$：

   * $E[z^2] = (2-1) E[z^0] = 1 \times 1 = \mathbf{1}$
   * $E[z^4] = (4-1) E[z^2] = 3 \times 1 = \mathbf{3}$
   * $E[z^6] = (6-1) E[z^4] = 5 \times 3 = \mathbf{15}$
   * $E[z^n] = (2n-1)!! $
   


## 第二部分：推廣至一般常態分佈以 $n=6$ 為例子 計算$E[x^6]$



根據 **【已知 1】** $x = \mu + \sigma z$ ， 利用 **【已知 3】 (二項式定理)** 將 $(\mu + \sigma z)^6$ 展開，係數為 `1, 6, 15, 20, 15, 6, 1` 。利用 **【已知 2】** 我們可以對每一項單獨取期望值，根據第一部分的結論所有含 $z$ 的奇數次方項期望值均為 $0$。




$$\begin{aligned}
E[x^6] =& E[(\mu + \sigma z)^6] \\ 
=& E[\mu^6 + 6\mu^5(\sigma z) + 15\mu^4(\sigma z)^2 + 20\mu^3(\sigma z)^3 + 15\mu^2(\sigma z)^4 + 6\mu(\sigma z)^5 + (\sigma z)^6 ]\\
=& \mu^6E[z^0] + 6\mu^5\sigma  E[z] + 5\mu^4\sigma^2  E[z^2] + 20\mu^3\sigma^3  E[z^3] + 15\mu^2\sigma^4  E[z^4] + 6\mu\sigma^5  E[z^5] + \sigma^6  E[z^6]\\
=& \mu^6 + 0 + 15\mu^4\sigma^2 + 0 + 45\mu^2\sigma^4 + 0 + 15\sigma^6 \\
=& \mu^6 + 15\mu^4\sigma^2 + 45\mu^2\sigma^4 + 15\sigma^6
\end{aligned}$$



## 程式實做檢驗

```python
import numpy as np
import matplotlib.pyplot as plt
import math

# ---------------------------------------------------------
# 1. 準備工作：定義雙階乘與理論公式函數
# ---------------------------------------------------------

def double_factorial(n):
    """
    計算雙階乘 n!!
    例如: 5!! = 5 * 3 * 1 = 15
    定義: (-1)!! = 1 (為了配合公式中 k=0 的情況)
    """
    if n <= 0:
        return 1
    result = 1
    for i in range(n, 0, -2):
        result *= i
    return result

def theoretical_moment(mu, sigma, n):
    """
    使用二項式展開與標準常態動差的通式計算 E[x^n]
    公式: Sum_{k=0 to n//2} C(n, 2k) * mu^(n-2k) * sigma^(2k) * (2k-1)!!
    """
    total = 0
    # k 代表我們選擇了幾個 z (z必須成對出現，所以是 2k)
    for k in range(0, n // 2 + 1):
        # 組合數 C(n, 2k)
        comb = math.comb(n, 2 * k)
        
        # mu 的項次
        mu_term = mu ** (n - 2 * k)
        
        # sigma 的項次
        sigma_term = sigma ** (2 * k)
        
        # 標準常態分佈的動差 E[z^(2k)] = (2k-1)!!
        z_moment = double_factorial(2 * k - 1)
        
        total += comb * mu_term * sigma_term * z_moment
        
    return total

# ---------------------------------------------------------
# 2. 設定實驗參數與生成數據
# ---------------------------------------------------------

# 為了讓結果好觀察，我們限制 mu 和 sigma 在一個合理範圍
# 如果 mu 太大，n=10 時數值會大到溢位或難以畫圖
np.random.seed(42) # 固定種子以便重現 (你可以註解掉這行來體驗完全隨機)
mu_real = np.random.uniform(0.5, 2.0)  # 隨機 mu
sigma_real = np.random.uniform(0.5, 1.5) # 隨機 sigma
n_range = range(1, 11)

def calculation_moment (sample_size=1000):
    sample_size = sample_size

    print(f"--- 實驗設定 ---")
    print(f"真實 Mu    = {mu_real:.4f}")
    print(f"真實 Sigma = {sigma_real:.4f}")
    print(f"樣本數 N   = {sample_size}")
    print(f"----------------")

    # 生成 1000 筆隨機數據
    data = np.random.normal(mu_real, sigma_real, sample_size)

    # ---------------------------------------------------------
    # 3. 計算 n=1 到 10 的理論值與樣本平均值
    # ---------------------------------------------------------

    theo_values = []
    emp_values = []
    errors = []

    print(f"{'n':<3} | {'理論值 E[x^n]':<15} | {'樣本平均 mean(x^n)':<15} | {'相對誤差 %':<10}")
    print("-" * 55)

    for n in n_range:
        # 理論值 (使用公式)
        theo = theoretical_moment(mu_real, sigma_real, n)
        
        # 樣本值 (直接計算數據的 n 次方，然後取平均)
        # 這就是蒙地卡羅模擬 (Monte Carlo Simulation) 的精神
        emp = np.mean(data ** n)
        
        # 計算相對誤差 (避免分母為0)
        if abs(theo) > 1e-9:
            error_pct = abs(theo - emp) / abs(theo) * 100
        else:
            error_pct = abs(theo - emp) # 如果理論值是0，就看絕對誤差
            
        theo_values.append(theo)
        emp_values.append(emp)
        errors.append(error_pct)
        
        print(f"{n:<3} | {theo:<15.4f} | {emp:<15.4f} | {error_pct:<9.2f}%")
        
    return theo_values , emp_values , errors

theo_values , emp_values , errors = calculation_moment(1000)
theo_values1w , emp_values1w , errors1w = calculation_moment(10000)
# ---------------------------------------------------------
# 4. 畫圖呈現
# ---------------------------------------------------------

fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 8))

# 上圖：數值比較 (使用 Log Scale 因為 n=10 時數值會非常巨大)
width = 0.35
x_idx = np.arange(len(n_range)) + 1

ax1.bar(x_idx - width/2*1.5, theo_values, width/2, label='Theoretical (Formula) 1k', alpha=0.7)
ax1.bar(x_idx - width/2*0.5, emp_values, width/2, label='Empirical (Simulation) 1k', alpha=0.7)
ax1.bar(x_idx + width/2*0.5, theo_values1w, width/2, label='Theoretical (Formula) 1w', alpha=0.7)
ax1.bar(x_idx + width/2*1.5, emp_values1w, width/2, label='Empirical (Simulation) 1w', alpha=0.7)

ax1.set_yscale('symlog') # 使用對稱對數座標，可以處理負值且壓縮巨大數值
ax1.set_title(f'Comparison of Moments ( $\mu$={mu_real:.2f}, $\sigma$={sigma_real:.2f})')
ax1.set_ylabel('Value (Log Scale)')
ax1.set_xticks(x_idx)
ax1.legend()
ax1.grid(True, which="both", ls="-", alpha=0.2)

# 下圖：相對誤差百分比

ax2.plot(x_idx, errors, marker='o', color='red', linestyle='-', linewidth=2,label='1k')
ax2.plot(x_idx, errors1w, marker='o', color='blue', linestyle='-', linewidth=2,label='1w')
ax2.set_title('Relative Error (%) vs Order n')
ax2.set_xlabel('Order (n)')
ax2.set_ylabel('Error %')
ax2.set_xticks(x_idx)
ax2.grid(True, alpha=0.5)

# 在誤差點上標註數字
for i, txt in enumerate(errors):
    ax2.annotate(f"{txt:.1f}%", (x_idx[i], errors[i]), xytext=(0, 5), textcoords='offset points', ha='center')
for i, txt in enumerate(errors1w):
    ax2.annotate(f"{txt:.1f}%", (x_idx[i], errors1w[i]), xytext=(0, 5), textcoords='offset points', ha='center')
ax2.legend()
plt.tight_layout()
plt.show()
```

* **為什麼 $n$ 越大，誤差通常越大？ (觀察圖表)**
  
  當你執行程式後看第二張圖的紅線(1000個樣本)你通常會發現 $n$ (moment)越高誤差百分比波動越大，這是統計學上很重要的現象：
  
  * 高次方的放大效應：當 $n=10$ 時，數據點裡只要有幾個稍微偏離平均值的「離群值」(Outlier)，經過 10 次方運算後（例如 $3^{10} \approx 59000$），它們對平均值的影響會變得非常巨大。
  * 樣本數限制：當我們只有 1000 筆數據對於估計 $E[x]$ (mean) 來說 1000 筆很夠了，但對於估計 $E[x^{10}]$ (Tenth moment) ，1000 筆數據往往不足以捕捉到分佈極端尾部的行為導致樣本變異數非常高。
 

  你可以試著把 sample_size 改成 100000，你會發現紅線（誤差）會瞬間掉下來，這就是大數法則 (Law of Large Numbers) 的威力 owob
