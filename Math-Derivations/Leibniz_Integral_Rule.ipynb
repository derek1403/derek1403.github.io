# Leibniz Integral Rule 萊布尼茲積分法則

## 證明目標

$$\frac{d}{dt} \int_{a(t)}^{b(t)} f(x,t) \, dx = \int_{a(t)}^{b(t)} \frac{\partial f(x,t)}{\partial t} \, dx + f(b(t),t) \frac{d b(t)}{dt} - f(a(t),t) \frac{d a(t)}{dt}$$



## 假設與已知 (Assumptions & Preliminaries)

* **【已知 1】標準常態分佈變數變換**  任何常態分佈 $x \sim \mathcal{N}(\mu, \sigma^2)$ 都可以寫成標準常態分佈 $z \sim \mathcal{N}(0, 1)$ 的線性變換：
  
  $$x = \mu + \sigma z$$
  
  其中 $z$ 的機率密度函數 (PDF) 為 $f(z) = \frac{1}{\sqrt{2\pi}} e^{-\frac{z^2}{2}}$。

* **【已知 2】期望值的線性性質** 對於常數 $a, b$ 與隨機變數 $X, Y$：
  
  $$E[aX + bY] = aE[X] + bE[Y]$$

* **【已知 3】二項式定理 (Binomial Theorem)** 
  
  $$(a+b)^n = \sum_{k=0}^{n} \binom{n}{k} a^{n-k} b^k$$
  
  舉例來說當 $n=6$ 時，巴斯卡三角形係數為：`1, 6, 15, 20, 15, 6, 1`

* **【已知 4】積分邊界條件收斂** 在Integration by Parts運算中，我們知道指數衰減函數 $e^{-\frac{x^2}{2}}$ 在無窮遠處的衰減速度快於任何多項式增長：
  
  $$\lim_{x \to \pm\infty} x^k e^{-\frac{x^2}{2}} = 0 \quad (\text{對於任意正整數 } k)$$
  

## **證明:**
你的原始筆記推導稍微有些跳躍。在嚴格數學上，我們透過定義一個多變數函數 $I(u, v, t) = \int_{u}^{v} f(x,t) \, dx$ 來證明。

$$\begin{gather*}
\text{由多變數連鎖律（Chain Rule）可知：} \\
\frac{d}{dt} I(a(t), b(t), t) = \frac{\partial I}{\partial t} + \frac{\partial I}{\partial b} \frac{db}{dt} + \frac{\partial I}{\partial a} \frac{da}{dt}
\end{gather*}$$

針對右邊三項分別計算：

1. **內部時間變化:** 固定邊界 $a, b$，對 $t$ 偏微分 $\Rightarrow \frac{\partial I}{\partial t} = \int_{a(t)}^{b(t)} \frac{\partial f(x,t)}{\partial t} \, dx$
2. **上限邊界變化:** 根據微積分基本定理 $\Rightarrow \frac{\partial I}{\partial b} = f(b(t),t)$
3. **下限邊界變化:** 根據微積分基本定理（下限帶負號）$\Rightarrow \frac{\partial I}{\partial a} = -f(a(t),t)$
