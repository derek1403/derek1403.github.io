# Hydrostatic Balance for Perturbations (擾動場的流體靜力平衡)



## 證明目標:

$$\frac{\partial\phi}{\partial z}=\frac{g}{\theta_{0}}\theta$$

* 註：為了符合大氣動力學的慣例，此處的 $\theta$ 代表的是**位溫的微小擾動量**（即下文推導中的 $\theta'$）。這是一個描述 **「垂直的擾動氣壓梯度力」** 與 **「浮力」** 之間互相平衡的經典關係式



## 假設與已知 (Assumptions & Preliminaries)

* **【已知 1】 圓柱座標的 $\hat{z}$ 垂直動量方程式：** $$\frac{Dw}{Dt} = -\frac{1}{\rho} \frac{\partial P}{\partial z} - g + \text{RHS}_{\hat{z}}$$

* **【假設 1】 靜力平衡假設 (Hydrostatic Assumption)：** $$\frac{Dw}{Dt} \ll g \quad \text{且} \quad \text{RHS}_{\hat{z}} = 0$$

  * 對於大尺度或中尺度系統（如颱風的環境場），垂直加速度與未解析的小尺度外力極小，可以忽略。
  * **尺度分析**：
    * 垂直風速 (一般對流或主環流)：$W \sim 1 \text{ m/s}$
    * 垂直高度尺度 (對流層)：$H \sim 10^4 \text{ m}$
    * 垂直加速度：$\frac{Dw}{Dt} \sim \frac{W^2}{H} \sim \frac{1^2}{10000} = \mathbf{10^{-4}} \text{ m/s}^2$
    * 重力加速度：$g \sim \mathbf{9.8} \text{ m/s}^2$
  * 垂直加速度的量級 ($10^{-4}$) 遠小於重力 ($10^1$)，因此將 $\frac{Dw}{Dt}$ 視為 0。

* **【已知 2】 雷諾分解 (Reynolds Decomposition)：** 我們將變數拆分為「只隨高度變化的背景平均態 (下標 0)」與「微小的擾動態 (上標 ' )」
  
  * 氣壓：$P = P_0(z) + P'$
  * 密度：$\rho = \rho_0(z) + \rho'$
  * 溫度：$T = T_0(z) + T'$
  * 位溫：$\Theta = \theta_0(z) + \theta'$

* **【假設 2】 背景場靜力平衡 (Background Hydrostatic Balance)：** 假設背景大氣本身是完美平衡的。
  
  $$\frac{\partial P_0}{\partial z} = -\rho_0 g$$

* **【已知 3】 泰勒展開近似 (Taylor Series Approximation)：** 對於微小擾動 $\rho' \ll \rho_0$，我們可以使用一階泰勒展開來處理密度的倒數：
  
  $$\begin{gather*}
  \frac{1}{\rho} &\overset{\text{已知 2}}{=}& \frac{1}{\rho_0 + \rho'} \\
  &=& \frac{1}{\rho_0 \left(1 + \frac{\rho'}{\rho_0}\right)} \\
  &\approx & \frac{1}{\rho_0} \left( 1 - \frac{\rho'}{\rho_0} \right)
  \end{gather*}$$

* **【已知 4】 狀態方程式 求 $\frac{dT}{T_0}$：**
  
  $$\begin{gather*}
  P &=& \rho R_d T \\
  \ln P &=& \ln(\rho R_d T)\\
  \ln P &=& \ln \rho + \ln R_d + \ln T \\
  d[\ln P] &=& d[\ln \rho] + d[\ln R_d] + d[\ln T] \\
  \frac{dP}{P} &=& \frac{d\rho}{\rho} + 0 + \frac{dT}{T}\\
  \frac{dP}{P} &=& \frac{d\rho}{\rho} + \frac{dT}{T}\\
  \frac{dT}{T} &=& \frac{dP}{P} - \frac{d\rho}{\rho}\\
  \end{gather*}$$ 


* **【已知 5】 位溫定義找和 $\frac{\rho'}{\rho_0}$ 的關係：**

  $$\begin{gather*}
  \Theta &\overset{\text{def}}{=}& T \left(\frac{P_s}{P}\right)^{R_d/c_p}\\
  \ln \Theta &=& \ln(T \left(\frac{P_s}{P}\right)^{R_d/c_p})\\
  \ln \Theta &=& \ln T + \ln \left(\frac{P_s}{P}\right)^{R_d/c_p} \\
  \ln \Theta &=& \ln T  + \frac{R_d}{c_p}\ln \left(\frac{P_s}{P}\right)  \\
  \ln \Theta &=& \ln T  - \frac{R_d}{c_p}\ln \left(\frac{P}{P_s}\right) \\
  d[\ln \Theta] &=& d[\ln T]  - d[\frac{R_d}{c_p}\ln \left(\frac{P}{P_s}\right)] \\
  \frac{d\Theta}{\Theta} &=& \frac{dT}{T}  - \frac{R_d}{c_p} \frac{d[\frac{P}{P_s}]}{\frac{P}{P_s}} \\
  \frac{d\Theta}{\Theta} &=& \frac{dT}{T}  - \frac{R_d}{c_p} \frac{dP}{P} \\
  \frac{d\Theta}{\Theta} &\overset{\text{已知 4}}{=}& \frac{dP}{P} - \frac{d\rho}{\rho}  - \frac{R_d}{c_p} \frac{dP}{P} \\
  \frac{d\Theta}{\Theta} &=& - \frac{d\rho}{\rho}  \left( 1-  \frac{R_d}{c_p} \right) \frac{dP}{P} \\
  \frac{\theta'}{\theta_0} &\overset{\text{已知 2}}{=}& - \frac{\rho'}{\rho_0}  + \left( 1-  \frac{R_d}{c_p} \right) \frac{P'}{P_0} \\
  \end{gather*}$$ 





* **【假設 3】 布氏近似 / 尺度分析 (Boussinesq Approximation)：**
  
  $$\frac{\rho'}{\rho_0} \approx -\frac{\theta'}{\theta_0}$$

  * 大氣中，壓力的相對擾動量通常遠小於溫度/位溫的相對擾動量。
  * **尺度分析**：
    * 背景壓力：$P_0 \sim 10^5 \text{ Pa}$
    * 擾動壓力：$P' \sim 100 \text{ Pa}$ ($1 \text{ hPa}$) $\implies \frac{P'}{P_0} \sim \mathbf{10^{-3}}$
    * 背景位溫：$\theta_0 \sim 300 \text{ K}$
    * 擾動位溫：$\theta' \sim 3 \text{ K}$ $\implies \frac{\theta'}{\theta_0} \sim \mathbf{10^{-2}}$
    * 乾空氣的氣體常數：$R_d \approx 287 \text{J}\cdot \text{kg}^{-1}\cdot \text{K}^{-1} $
    * 乾空氣的定壓比熱：$C_p \approx 1004 \text{J}\cdot \text{kg}^{-1}\cdot \text{K}^{-1}$
    * $R_d / C_p \approx 0.286$
  * 因為 $\frac{P'}{P_0} \ll \frac{\theta'}{\theta_0}$，【已知 5】中的壓力項可被忽略。這表示密度的改變主要由 **位溫** 決定

* **【定義 1】 運動學壓力擾動 $\phi$ (Kinematic Pressure Perturbation)：** $$\phi \overset{\text{def}}{=} \frac{P'}{\rho_0}$$

  * 註：在布氏近似或非彈性近似中，通常視 $\rho_0$ 的垂直變化遠小於 $P'$ 的垂直變化，因此 $\frac{\partial \phi}{\partial z} \approx \frac{1}{\rho_0}\frac{\partial P'}{\partial z}$。



## 證明:

$$\begin{gather*}
\frac{Dw}{Dt} &\overset{\text{已知 1}}{=}& -\frac{1}{\rho} \frac{\partial P}{\partial z} - g + \text{RHS}_{\hat{z}} \\
0 &\overset{\text{假設 1}}{=}& -\frac{1}{\rho} \frac{\partial P}{\partial z} - g \\
0 &\overset{\text{已知 2}}{=}& -\frac{1}{\rho_0 + \rho'} \frac{\partial}{\partial z} \left[ P_0 + P' \right] - g \\
0 &\overset{\text{已知 3}}{=}& -\frac{1}{\rho_0} \left( 1 - \frac{\rho'}{\rho_0} \right) \left[ \frac{\partial P_0}{\partial z} + \frac{\partial P'}{\partial z} \right] - g \\
0 &=& -\frac{1}{\rho_0}\frac{\partial P_0}{\partial z} - \frac{1}{\rho_0}\frac{\partial P'}{\partial z} + \frac{\rho'}{\rho_0^2}\frac{\partial P_0}{\partial z} + \underbrace{\frac{\rho'}{\rho_0^2}\frac{\partial P'}{\partial z}}_{\text{高階項，忽略}} - g \\
0 &\overset{\text{假設 2}}{=}& -\frac{1}{\rho_0}(-\rho_0 g) - \frac{1}{\rho_0}\frac{\partial P'}{\partial z} + \frac{\rho'}{\rho_0^2}(-\rho_0 g) - g \\
0 &=& g - \frac{1}{\rho_0}\frac{\partial P'}{\partial z} - g\frac{\rho'}{\rho_0} - g \\
0 &=& -\frac{1}{\rho_0}\frac{\partial P'}{\partial z} - g\frac{\rho'}{\rho_0} \\
0 &\overset{\text{假設 3}}{=}& -\frac{1}{\rho_0}\frac{\partial P'}{\partial z} - g\left(-\frac{\theta'}{\theta_0}\right) \\
\frac{1}{\rho_0}\frac{\partial P'}{\partial z} &=& g\frac{\theta'}{\theta_0} \\
\frac{\partial \phi}{\partial z} &\overset{\text{定義 1}}{=}& g\frac{\theta'}{\theta_0} \\
\frac{\partial\phi}{\partial z} &=& \frac{g}{\theta_{0}}\theta \quad \text{(令符號 } \theta' \text{ 為 } \theta \text{)}
\end{gather*}$$

## 物理解釋

### 步驟一：方程式的右邊 $\frac{g}{\theta_{0}}\theta'$ —— 浮力的來源

在流體力學中，這個項代表的就是**浮力 (Buoyancy)**。

* 當一團空氣比周圍環境還要溫暖時（位溫擾動 $\theta' > 0$），它的密度會變小。
* 根據阿基米德原理，密度比周圍小的物體會受到一個向上的浮力。
* 所以，**等式右邊的物理意義是：這團空氣因為溫度異常，產生了多大的淨浮力。** 若 $\theta' > 0$，浮力向上（正值）；若 $\theta' < 0$（冷空氣），浮力向下（負值）。

### 步驟二：方程式的左邊 $\frac{\partial\phi}{\partial z}$ —— 擾動氣壓的垂直變化

回憶一下 $\phi = \frac{P'}{\rho_0}$，所以 $\frac{\partial\phi}{\partial z}$ 其實就是在描述「擾動氣壓隨高度的變化率」。

* 我們知道，氣壓梯度力 (Pressure Gradient Force) 的方向是**從高壓指向低壓**。在垂直方向上，氣壓梯度力正比於 $-\frac{\partial\phi}{\partial z}$（注意那個負號）。
* 如果 $\frac{\partial\phi}{\partial z} > 0$，意思是「越往高空，擾動氣壓越大」。高空氣壓大、低空氣壓小，這會產生一個**向下的擾動氣壓梯度力**。
* 所以，**等式左邊的物理意義是：大氣必須要建立起一種特定的氣壓垂直分布，來對抗其他力。**

### 步驟三：等號 $=$ —— 大自然的妥協與平衡

為什麼它們會相等？因為我們假設了系統處於**靜力平衡（垂直加速度 $\frac{Dw}{Dt} \approx 0$）**。

這意味著這團空氣雖然很熱、很有向上衝的衝動（有浮力），但它**並沒有**瘋狂地向上加速。為什麼沒有？因為大自然產生了一個大小相等、方向相反的「擾動氣壓梯度力」把它硬生生壓住了。

### 💡 老師的直觀比喻：

**1. 基本比喻：潛水中的熱氣球**
想像你在水池底部（背景平衡態）放開一顆充滿熱氣的氣球（正位溫擾動 $\theta' > 0$）。氣球本來應該要往上衝（浮力），但如果我告訴你這顆氣球現在「懸浮在水中沒有加速往上」，那是為什麼？
因為我用手在氣球上方用力往下壓（建立了 $\frac{\partial\phi}{\partial z} > 0$ 的壓力場）。這隻手往下壓的力道，剛剛好抵消了氣球向上的浮力。這就是擾動場的流體靜力平衡。

**2. 進階大氣應用：為什麼颱風中心是強烈低氣壓？**
這個方程式完美解釋了**颱風的暖心結構 (Warm Core)** 與地面低壓的關係！

* **右邊 (浮力)：** 颱風的眼牆內部有大量的潛熱釋放，導致颱風中心從低空到高空都充滿了非常溫暖的空氣（$\theta' > 0$ 極大）。這產生了極強的向上浮力。
* **左邊 (氣壓場反應)：** 為了維持平衡，颱風中心的擾動氣壓必須滿足 $\frac{\partial\phi}{\partial z} > 0$（即 $\frac{g}{\theta_{0}}\theta'$ 是正的）。
* **物理推論：** $\frac{\partial\phi}{\partial z} > 0$ 代表「擾動氣壓由下往上遞增」。我們知道在大氣最頂層（平流層頂），擾動氣壓幾乎為零（$\phi_{\text{top}} \approx 0$）。如果氣壓是**由下往上遞增到零**，這代表什麼？
* **結論：** 這代表底層的擾動氣壓必須是**極大的負值 ($\phi_{\text{surface}} \ll 0$)**！

這就是為什麼當大氣中存在一根巨大的「暖柱」時，地面就必然會對應一個「強烈低氣壓」。這條短短的公式 $\frac{\partial\phi}{\partial z} = \frac{g}{\theta_{0}}\theta$，用最精煉的數學，把熱力學（溫度擾動）與動力學（氣壓場建立）完美地綁在一起了！
