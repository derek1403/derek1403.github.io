
#  位渦 $PV$ 與浮力位渦 $PV_b$ 的關係 (Relationship between Ertel Potential Vorticity and Buoyancy Potential Vorticity)


## 證明目標


$$PV = \left(\frac{\theta_0}{\rho_0 g}\right) PV_b$$


## 假設與已知 (Assumptions & Preliminaries)


* **【已知 1】 Ertel Potential Vorticity 定義：**

    $$PV \overset{\text{def}}{=} \frac{1}{\rho}\boldsymbol{\eta}_a \cdot \nabla\theta$$

    * $\boldsymbol{\eta}_a$ 是絕對渦度向量
    * $\rho$ 是密度
    * $\theta$ 是位溫

* **【已知 2】 浮力與位溫的關係：**

    我們之前定義過浮力 $b$ 為：
    $$b = \frac{g}{\theta_0}\theta$$
    其中 $g$ 和 $\theta_0$ 都是常數。

* **【假設 1】 布氏近似 (Boussinesq Approximation)：**

    在布氏近似下，我們假設流體的密度變化非常微小，所以在作為「慣性質量」時（也就是出現在分母時），我們可以用常數參考密度 $\rho_0$ 來取代真實密度 $\rho$。

* **【定義 1】 Buoyancy Potential Vorticity：**

    $$PV \overset{\text{def}}{=} \left(\frac{\theta_0}{\rho_0 g}\right) PV_b$$




## 證明:


$$\begin{gather*}
&PV &\overset{\text{已知 1}}{=}& \frac{1}{\rho} \boldsymbol{\eta}_a \cdot \nabla \theta\\
&PV &\overset{\text{已知 2}}{=}& \frac{1}{\rho} \boldsymbol{\eta}_a \cdot \left(\frac{\theta_0}{g}\nabla b\right)\\
&PV &\overset{\text{假設 1}}{\approx}& \frac{1}{\rho_0} \boldsymbol{\eta}_a \cdot \left(\frac{\theta_0}{g}\nabla b\right)\\
&PV &=&  \left(\frac{\theta_0}{\rho_0 g}\right) (\boldsymbol{\eta}_a \cdot \nabla b)\\
PV_b\overset{\text{定義 1}}{=}&PV \left(\frac{\rho_0 g}{\theta_0}\right)&=&   \boldsymbol{\eta}_a \cdot \nabla b\\
\end{gather*}$$

### 💡 物理直覺與類比

你可以把位渦想像成你的「財富價值」。
* **經典 $PV$** 就像是你帳戶裡有幾塊錢 **美金 (USD)**。它的單位包含了嚴格的密度 $\rho$ 和位溫 $\theta$。
* **浮力位渦 $PV_b$** 就像是你帳戶裡有幾塊錢 **台幣 (TWD)**。

那串常數 $\left(\frac{\theta_0}{\rho_0 g}\right)$ 就是**「固定匯率」**！

因為在我們探討颱風次環流的系統裡，我們已經把所有的變數都換成了「台幣（浮力 $b$）」在計算（例如垂直穩定度 $A = \frac{\partial b}{\partial z}$）。為了讓計算順暢，我們直接用 $PV_b$（台幣計算的位渦）來衡量系統的穩定度。

[cite_start]數值上它們不一樣大，但它們衡量的是**一模一樣的物理性質（流體對稱擾動的剛性）**。只要 $PV_b$ 是守恆的，經典的 $PV$ 就一定也是守恆的！講義上直接寫 $AC - B^2 = PV$ [cite: 100, 117]，其實就是在這個布氏近似與變數代換的框架下，省略了常數轉換的結果。
