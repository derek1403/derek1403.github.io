# Terminal Velocity

## **物理意義**

**終端速度（Terminal Velocity）** 是指物體在流體（如空氣或水）中下落(上浮)時，當向下的重力與向上的拖曳力（阻力、浮力） 達到平衡，**合力為零從而停止加速，並維持恆定不變的最大速度**

**拖曳力（Drag Force）** 其實是一個總稱，它包含了流體對物體的所有阻力，例如**黏滯力（Viscous Force）**、**慣性力（Inertial Force，又稱形狀阻力）**

**[雷諾數$Re$](./Reynolds_number.ipynb)** 是由慣性力和黏滯力的比值所形成的無因次參數，當 $Re \gg 1$ (雷諾數很高)的時候，物體受到的力主要由慣性力主導，當 $Re \ll 1$  (雷諾數很低)的時候，物體受到的力主要由黏滯力主導。




## **從空中掉落的雨滴的終端速度 (Terminal velocity of air)**

在大氣中，因為空氣密度極小，我們通常忽略空氣給雨滴的浮力。

### 方程式

**力學平衡方程式：** $\text{重力} = \text{拖曳力}$

**1. 驅動力：重力 ($F_g$)**

雨滴的質量是體積乘上密度

$$F_g = \frac{4}{3}\pi r^3 \rho_{\text{drop}} g$$

* $r$ 是雨滴半徑(**特徵尺度**)，單位 $[\text{m}]$
* $\rho_{\text{drop}}$ 是水滴密度，單位 $[\text{kg}^2 \cdot \text{m}^{-3}]$
* $g$ 是重力加速度，$9.81[\text{m}^2 \cdot \text{s}^{-2}]$



**2. 阻力：拖曳力 ($F_\text{d}$)**

**低 Re（黏滯力主宰）**：這時阻力與速度的一次方成正比斯托克斯定律（Stokes' law）

$$F_{\text{d, viscous}} = 6\pi \mu_{\text{air}} v r$$



* $\mu_{\text{air}}$ 是空氣動態黏滯係數，單位 $[\text{kg}^2 \cdot \text{m}^{-1}\cdot \text{s}^{-1}] = [\text{Pa} \cdot \text{s}]$
* $v$ 是終端速度(**特徵速度**)，單位 $[\text{m} \cdot \text{s}^{-1}]$
* $r$ 是雨滴半徑(**特徵尺度**)，單位 $[\text{m}]$

**高 Re（慣性力主宰）**：這時阻力與速度的平方、以及迎風截面積（$\pi r^2$）成正比（牛頓阻力）。

$$F_{\text{d, inertial}} = \frac{1}{2} C_D \rho_{\text{air}} v^2 (\pi r^2)$$


* $C_D$ 是阻力係數，在高 Re 時可視為常數，無因次量
* $\rho_{\text{air}}$ 是空氣密度，單位 $[\text{kg}^2 \cdot \text{m}^{-3}]$
* $v$ 是終端速度(**特徵速度**)，單位 $[\text{m} \cdot \text{s}^{-1}]$
* $r$ 是雨滴半徑(**特徵尺度**)，單位 $[\text{m}]$


### 平衡計算

**低 Re（黏滯力主宰）** 情況：

$$\begin{gather*}
F_g &=& F_\text{d} \\ 
\frac{4}{3}\pi r^3 \rho_{\text{drop}} g &=& 6\pi \mu_{\text{air}} v r \\
v &=& \frac{\frac{4}{3}\pi r^3 \rho_{\text{drop}} g }{6\pi \mu_{\text{air}}r} \\
v &=& \frac{2}{9} \frac{g}{\mu_{\text{air}}} \rho_{\text{drop}} r^2 \\
v &\sim& r^2
\end{gather*}$$

**高 Re（慣性力主宰）** 情況：

$$\begin{gather*}
F_g &=& F_\text{d} \\ 
\frac{4}{3}\pi r^3 \rho_{\text{drop}} g &=& \frac{1}{2} C_D \rho_{\text{air}} v^2 (\pi r^2) \\
v &=& \sqrt{\frac{\frac{4}{3}\pi r^3 \rho_{\text{drop}} g }{\frac{1}{2} C_D \rho_{\text{air}} (\pi r^2)} }\\
v &=& \sqrt{\frac{8}{3}\frac{\rho_{\text{drop}}}{\rho_{\text{air}}}\frac{g}{C_D}r}\\
v &\sim& r^{0.5}
\end{gather*}$$


## **從海水中上浮的氣泡的終端速度 (Terminal velocity of water)**


在海洋中，氣泡是往上跑的，這時「浮力」遠大於氣泡本身的「重力」，所以驅動力變成了向上的「淨浮力」。


### 方程式

**力學平衡方程式：** $\text{淨浮力} = \text{拖曳力}$

**1. 驅動力：淨浮力 ($F_{\text{net}}$)**

排開海水的重量減去氣體本身的重量

$$F_{\text{net}} = F_{\text{buoyancy}} - F_g = \frac{4}{3}\pi r^3 \rho_{\text{water}} g - \frac{4}{3}\pi r^3 \rho_{\text{gas}} g = \frac{4}{3}\pi r^3 (\rho_{\text{water}} - \rho_{\text{gas}}) g$$

* $r$ 是氣塊半徑(**特徵尺度**)，單位 $[\text{m}]$
* $\rho_{\text{water}}$ 是海水密度，單位 $[\text{kg}^2 \cdot \text{m}^{-3}]$
* $\rho_{\text{gas}}$ 是氣塊密度，單位 $[\text{kg}^2 \cdot \text{m}^{-3}]$
* $g$ 是重力加速度，$9.81[\text{m}^2 \cdot \text{s}^{-2}]$



**2. 阻力：拖曳力 ($F_\text{d}$)**

**低 Re（黏滯力主宰）**：

這裡有一個隱藏的流體力學細節！水流過「固體」表面（如石頭）和流過「氣體」表面（如氣泡）的摩擦條件不同，氣泡內部會產生氣流循環減少了表面的摩擦，所以對於極純淨水中的小氣泡，係數會從 $6$ 變成 $4$，如果是充滿雜質的真實海水，雜質會包覆氣泡，讓它表現得像固體，這時用 $6\pi$ 也可以。

$$F_{\text{d, viscous}}= 4\pi \mu_{\text{water}} v r$$

* $\mu_{\text{water}}$ 是水動態黏滯係數，單位 $[\text{kg}^2 \cdot \text{m}^{-1}\cdot \text{s}^{-1}] = [\text{Pa} \cdot \text{s}]$
* $v$ 是終端速度(**特徵速度**)，單位 $[\text{m} \cdot \text{s}^{-1}]$
* $r$ 是雨滴半徑(**特徵尺度**)，單位 $[\text{m}]$


**高 Re（慣性力主宰）**：

形式與大氣中一樣，只是流體密度變成了水。

$$F_{\text{d, inertial}} = \frac{1}{2} C_D \rho_{\text{water}} v^2 (\pi r^2)$$

* $C_D$ 是阻力係數，在高 Re 時可視為常數，無因次量
* $\rho_{\text{water}}$ 是水密度，單位 $[\text{kg}^2 \cdot \text{m}^{-3}]$
* $v$ 是終端速度(**特徵速度**)，單位 $[\text{m} \cdot \text{s}^{-1}]$
* $r$ 是空氣半徑(**特徵尺度**)，單位 $[\text{m}]$



### 平衡計算

**低 Re（黏滯力主宰）** 情況：

$$\begin{gather*}
F_\text{net} &=& F_\text{d} \\ 
\frac{4}{3}\pi r^3 (\rho_{\text{water}} - \rho_{\text{gas}}) g &=& 4\pi \mu_{\text{water}} v r \\
v &=& \frac{\frac{4}{3}\pi r^3 (\rho_{\text{water}} - \rho_{\text{gas}}) g}{4\pi \mu_{\text{water}}r} \\
v &=& \frac{1}{3} \frac{g}{\mu_{\text{water}}} (\rho_{\text{water}} - \rho_{\text{gas}}) r^2 \\
v &\sim& r^2
\end{gather*}$$

**高 Re（慣性力主宰）** 情況：

$$\begin{gather*}
F_g &=& F_\text{d} \\ 
\frac{4}{3}\pi r^3 (\rho_{\text{water}} - \rho_{\text{gas}}) g&=& \frac{1}{2} C_D \rho_{\text{water}} v^2 (\pi r^2) \\
v &=& \sqrt{\frac{\frac{4}{3}\pi r^3 (\rho_{\text{water}} - \rho_{\text{gas}}) g }{\frac{1}{2} C_D \rho_{\text{water}} (\pi r^2)} }\\
v &=& \sqrt{\frac{8}{3}\frac{(\rho_{\text{water}} - \rho_{\text{gas}})}{\rho_{\text{water}}}\frac{g}{C_D}r}\\
v &\sim& r^{0.5}
\end{gather*}$$
