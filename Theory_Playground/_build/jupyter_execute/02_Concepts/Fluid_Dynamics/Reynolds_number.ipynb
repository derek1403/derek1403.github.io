# Reynolds Number $Re$

## 定義 **雷諾數 $Re$**

$Re$ 是由慣性力和黏滯力的比值所形成的無因次參數，當 $Re \gg 1$ (雷諾數很高)的時候，物體受到的力主要由慣性力主導，當 $Re \ll 1$  (雷諾數很低)的時候，物體受到的力主要由黏滯力主導。


$$\begin{gather*}
Re & \overset{\text{def}}{=}& \frac{\text{慣性力}}{\text{黏滯力}} \\
&=& \frac{\rho u r}{\mu} \\
&=& \frac{ u r}{\nu} \\
&=& \frac{L^2 / \nu}{L/U} \\
&=& \frac{T_{\text{viscous}}}{T_{\text{inertia}}} &=&  \frac{\text{物體感受到黏滯力時間尺度}}{\text{物體感受到慣性力時間尺度}}
\end{gather*}$$


* $U$ 是**特徵速度**，單位 $[\text{m} \cdot \text{s}^{-1}]$
* $L$ 是**特徵長度**，單位 $[\text{m}]$
  
  是**速度場（Velocity Field）發生顯著變化的空間跨度**，例如:
  * 雨滴：速度從水滴表面的 $U$ 降到遠處的 0，跨度大約是 雨滴半徑。
  * 颱風：速度從眼牆的 $U_{max}$ 降到外圍的 0，跨度大約是 颱風半徑。
    * $\delta_{颱風} = \frac{15 \text{ km}}{500 \text{ km}} \approx 0.03 \ll 1$ ，垂直速度遠小於水平速度，主要受阻力在水平方向上
  * 颱風中的一個小氣塊：速度從氣塊表面的 $U$ 降到遠處的 0，跨度大約是 氣塊半徑。
  * 水管：速度從管中心的 $U_{max}$ 降到管壁的 0，跨度大約是 水管直徑。
  * 對流雲：地面被加熱空氣往上衝，一直衝到「對流層頂（大約 10~15 公里高）」遇到穩定的平流層被迫向外散開
    * $\delta_{對流雲} = \frac{10 \text{ km}}{10 \text{ km}} \approx 1$ (各向同性)，垂直速度大於水平速度，主要受阻力在垂直方向上
  
  通常**假設氣塊是球**就會用半徑$r$來代表特徵尺度

* $\nu$ 是**運動黏滯係數(Kinematic Viscosity)**，單位 $[\text{m}^2 \cdot \text{s}^{-1}]$
  * $\nu \overset{\text{def}}{=}\frac{\mu}{\rho}$
* $\mu$ 是**動態黏滯係數(Dynamic Viscosity)**，單位 $[\text{kg}^2 \cdot \text{m}^{-1}\cdot \text{s}^{-1}] = [\text{Pa} \cdot \text{s}]$
* $\rho$ 是**密度**，單位 $[\text{kg}^2 \cdot \text{m}^{-3}]$
  
  是被擾動的背景的密度例如:
  * 水中的氣泡： 流動的是水，所以 $\rho$ 帶入水的密度，$\mu$ 帶入水的黏滯係數。
  * 空氣中的雨滴： 流動的是空氣，所以 $\rho$ 帶入空氣密度，$\mu$ 帶入空氣黏滯係數。
  * 大氣中的氣塊（例如對流雲）： 流動的是周圍的環境空氣，所以 $\rho$ 和 $\mu$ 都是帶入大氣環境的值。



## 常見參數參考表


* **重力加速度 ($g$)：**  $9.81[\text{m}^2 \cdot \text{s}^{-2}]$
* **水的密度 ($\rho_{water}$)：**  $1000[\text{kg}^2 \cdot \text{m}^{-3}]$
  * 常溫常壓（約 20°C, 1 atm）下的近似值
* **空氣的密度 ($\rho_{air}$)：**  $1.2[\text{kg}^2 \cdot \text{m}^{-3}]$
  * 常溫常壓（約 20°C, 1 atm）下的近似值
* **水的動態黏滯係數 ($\mu_{water}$)：** $1.0 \times 10^{-3} [\text{kg}^2 \cdot \text{m}^{-1}\cdot \text{s}^{-1}]$
* **空氣的動態黏滯係數 ($\mu_{air}$)：** $1.8 \times 10^{-5} [\text{kg}^2 \cdot \text{m}^{-1}\cdot \text{s}^{-1}]$ 
* **空氣的運動黏滯係數 ($\nu_{air}$)：** $1.5 \times 10^{-5} [\text{m}^2 \cdot \text{s}^{-1}]$ 



## 多少算高 Re？多少算低 Re？

對於流過球體的流場，工程與物理上通常這樣界定：

### **低雷諾數（黏滯力主宰）：$Re < 1$**

此時流線非常平滑完全貼合球體稱為蠕流（Creeping flow），斯托克斯定律（Stokes law）完美適用，慣性力幾乎不存在，動量瞬間就被黏滯力擴散掉，物體只要一停止出力就會瞬間停在原地，完全沒有「滑行」這回事。

* **ex1:** 精子或大腸桿菌在水中游動（微觀視角）
  * **特徵長度 $L$：** 細菌長度約 $2 [\mu\text{m}] = 2 \times 10^{-6} [\text{m}]$
  * **特徵速度 $U$：** 游動速度約 $20 [\mu\text{m}\cdot \text{s}^{-1}] = 2 \times 10^{-5} [\text{m}\cdot \text{s}^{-1}]$
  * **水的運動黏滯係數 $\nu$：** $10^{-6} [\text{m}^2\cdot \text{s}^{-1}]$
  * **帶入公式：** $Re = \frac{2 \times 10^{-5} \times 2 \times 10^{-6}}{10^{-6}} = 0.00004 \ll 1$
  
  物理直覺： 對細菌來說，水根本不像水，水黏稠得像是柏油或濃稠的蜂蜜！在這麼低的雷諾數下，物理學家珀塞爾（E.M. Purcell）提出了一個著名的「扇貝定理（Scallop Theorem）」。如果你像扇貝一樣，只是簡單地「打開、合上、打開、合上」對稱地拍打水，在低雷諾數下你只會原地前後抖動，根本無法前進！（因為沒有慣性可以讓你在合上的瞬間滑行）。所以細菌必須演化出「螺旋狀的鞭毛」，像開瓶器一樣在水裡「鑽」著前進。

* **ex2:** 地球的地函對流（宏觀但極慢）
  * **特徵長度 $L$：** 地函厚度約 $3000 \text{ km} = 3 \times 10^6 \text{ m}$（極大！）
  * **特徵速度 $U$：** 板塊移動速度大約每年幾公分，算 $10^{-9} \text{ m/s}$（極慢！）
  * **地函的運動黏滯係數 $\nu$：** 大約高達 $10^{17} \text{ m}^2\text{/s}$（極度黏！）
  * **帶入公式：** $Re = \frac{10^{-9} \times 3 \times 10^6}{10^{17}} = 3 \times 10^{-20}$
  
  這是物理界能找到**最小的雷諾數之一**！所以地函對流是完美的層流，沒有任何亂流，幾億年來都維持著極度平滑的運動軌跡。



### **過渡區：$1 < Re < 1000$**

流體開始在物體後方產生穩定的漩渦，黏滯力無法完全壓制慣性，流場開始在物體後方「剝離（Separation）」產生**漩渦**，但這些漩渦還沒有碎裂成混亂的亂流，而是呈現極度規律的幾何排列。

* **ex 1:** 微風吹過電線發出「嗡嗡聲」（卡門渦街）

    當微風吹過一根細細的電線或天線時，我們來算算看：

    * **特徵長度 $L$：** 電線直徑約 $2 \text{ mm} = 0.002 \text{ m}$
    * **特徵速度 $U$：** 微風約 $1.5 \text{ m/s}$
    * **空氣運動黏滯係數 $\nu$：** $1.5 \times 10^{-5} \text{ m}^2\text{/s}$
    * **帶入公式：** $Re = \frac{1.5 \times 0.002}{1.5 \times 10^{-5}} = 200$

    在這個大約 $Re \approx 200$ 的過渡區，流體會在電線後方交替地產生一左一右的漩渦，這被稱為 **卡門渦街（Von Kármán vortex street）** ，這交替脫落的漩渦會對電線產生週期性的拉扯力，讓電線震動。如果震動頻率剛好落在人耳的聽覺範圍內，你就會聽到電線在風中「唱歌」！

    大氣中的例子
    
    [![](https://earthdata.nasa.gov/s3fs-public/styles/hds_large/public/2022-05/GuadalupeIsland_10Apr2022_Suomi_lg.jpeg?VersionId=SX5Oi4ofKSU0lQL3QLckmBtUxP4YxC7Q&itok=7o9k6ptr)](https://www.earthdata.nasa.gov/news/worldview-image-archive/swirls-clouds-off-guadalupe-island-baja-california-mexico)

* **ex 2:** 果蠅與蚊子的飛行

    這也是生物學家曾經困惑很久的謎題。

    * **特徵長度 $L$：** 昆蟲翅膀寬度約 $1 \text{ mm} = 10^{-3} \text{ m}$
    * **特徵速度 $U$：** 拍翅速度約 $1 \text{ m/s}$
    * **帶入公式：** $Re = \frac{1 \times 10^{-3}}{1.5 \times 10^{-5}} \approx 66$

    這剛好落在過渡區！早期的空氣動力學家如果用老鷹或飛機（高雷諾數）的公式去算，會發現蜜蜂和果蠅產生的升力根本不足以飛起來（著名的都市傳說：根據物理定律，大黃蜂不應該會飛）。
    後來才發現，昆蟲就是利用了**過渡區會產生穩定漩渦**的特性。牠們在翅膀前緣製造出一個「前緣渦流（Leading-Edge Vortex）」，這個漩渦沒有碎成亂流，而是黏在翅膀上，像一個低壓吸塵器一樣把昆蟲往上吸，提供了額外的巨大升力！


### **高雷諾數（慣性力主宰）：$Re > 1000$**

後方產生混亂的尾流（Wake），阻力係數 $C_D$ 趨近於常數（約 0.4 到 0.5）。這時就是 $v \sim r^{0.5}$ 成立的區間。

* **範例 1：對流雲（積雨雲的上升氣流）**

  * 特徵長度 $L$：對流雲的尺度大約 1 km 到 10 km，我們取 1000 m。
  * 特徵速度 $u$：強烈的上升氣流大約 10 m/s。
  * 代入公式： $Re = \frac{u L}{\nu_{air}} = \frac{10 \times 1000}{1.5 \times 10^{-5}} \approx 6.7 \times 10^8$

  這代表對流雲內部的流動是**極度高雷諾數**，慣性力遠大於黏滯力，這也是為什麼雲裡面的氣流總是呈現高度混亂的湍流（Turbulence）狀態。

* **範例 2：颱風（Typhoon / Hurricane）**

  * 特徵長度 $L$：颱風半徑大約 100 km，即 100,000 m。
  * 特徵速度 $u$：最大風速算 50 m/s。
  * 代入公式：$Re = \frac{u L}{\nu_{air}} = \frac{50 \times 100,000}{1.5 \times 10^{-5}} \approx 3.3 \times 10^{11}$

  這是一個天文數字！這說明在巨觀的天氣系統中，空氣分子的摩擦力（分子黏滯性）微乎其微，幾乎可以忽略不計。

>**進階概念延伸：**
>既然颱風的雷諾數這麼大，分子黏滯性根本不起作用，那為什麼颱風的風速不會無止盡增加？能量是如何耗散的？
>這就帶出了在數值天氣預報（如 FCNv2 或其他大氣模型）中一個極度重要的概念：**渦旋黏滯係數（Eddy Viscosity）**。
>
>在大尺度下，大氣不是靠「分子摩擦」來消耗能量，而是靠「大小不一的亂流漩渦」互相碰撞來傳遞和消耗動量。所以大氣科學家發明了「渦旋黏滯係數 $K_m$」來取代公式裡的 $\mu$。這個 $K_m$ 比分子的 $\mu$ 大了成千上萬倍！
>
>這就像是：在擁擠的夜市裡（大氣亂流），你無法前進不是因為空氣很黏（分子黏滯），而是因為你一直撞到其他走來走去的人（渦旋動量交換）。
