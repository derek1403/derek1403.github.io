# Boussinesq Approximation

## 核心假設：$\rho(x,y,z,t) \to \rho_0$ (完全是個常數！)

**數學操作：** Boussinesq 近似非常霸道，它宣稱流體的密度到處都一樣，完全是一個常數 $\rho_0$（例如 $1.2 \text{ kg/m}^3$）。唯一的例外（浮力項）： 如果密度到處都一樣，熱空氣怎麼上升？Boussinesq 規定，只有在計算垂直方向的「重力/浮力」時，才允許密度有一點點微小的擾動 $\rho'$，在其他所有的動量項、連續方程式裡，$\rho$ 通通被當作死硬的常數 $\rho_0$ 提出去消掉


。連續方程式會退化成最簡單的不可壓縮流：$\nabla \cdot \vec{V} = 0$（也就是 $\frac{\partial u}{\partial x} + \frac{\partial v}{\partial y} + \frac{\partial w}{\partial z} = 0$）。



$\frac{\partial \rho}{\partial t} = 0$

## 假設與已知 (Assumptions & Preliminaries)

* **【假設 1】 淺水假設：** 

  * 流體的垂向運動尺度比整個分析流域的垂向代表性尺度要小很多 **(不要有太大的垂直運動)**
  
* **【假設 2】 局部壓力梯度不改變慣性與質量的密度 ：** 

  * **白話翻譯：** 密度的微小變化 $\Delta \rho$，在算「推力 (慣性)」和「擠壓 (連續方程式)」的時候可以直接當作 0 忽略掉；只有在算「浮力」的時候才准出現。
  * **物理意義：** 
    * **算慣性 ($F=ma$) 時：** 推動這團空氣的難易度取決於總質量，$1.2$ $\text{kg} \cdot \text{m}^{-3}$ 跟 $1.21$ $\text{kg} \cdot \text{m}^{-3}$ 幾乎沒差別，所以慣性項裡的密度直接當作常數 $\rho_0$。
    * **算質量守恆時：** 把它當作完全壓不扁的水球，流進來多少就流出去多少。
    * **算浮力的時候：** 因為重力 $g$ ($9.8$ $\text{m} \cdot \text{s}^{-2}$) 非常強大，它會把這個微小的 $\Delta \rho$ 放大，所以很重要不能忽略
  * **比喻：** 想像一個體重 $100$ $\text{kg}$ 的壯漢，手裡拿著一顆 $10$ $\text{g}$ 的氦氣球。
    * 當我想從背後水平推壯漢（算慣性力 $F=ma$），這 $10$ $\text{g}$ 的氦氣球完全不影響我推壯漢的感覺
    * 在垂直方向上，這 $10$ $\text{g}$ 的氦氣球因為比周圍空氣輕，會產生向上的浮力。
    * 這就是 Boussinesq 近似的精髓：水平推的時候忽略微小差異，垂直飄的時候保留微小差異。

* **【假設 3】 低馬赫數假設：** 

  * **白話翻譯：** 流體移動得夠慢，慢到空氣有足夠的時間「讓開」，而不會被「撞扁」。
  * **物理意義：** 聲速（大約 $340$ $\text{m} \cdot \text{s}^{-1}$）其實就是空氣壓力波傳遞的速度。流體的速度 $V$ 除以聲速 $c$，就是馬赫數 ($M = V/c$)。當馬赫數遠小於 1 ($M \ll 1$) 時，動能轉換成壓力變化的比例極小，流體就來不及被壓縮 。
  * **比喻：** 想像你在一個擁擠的演唱會人群中。
    * 如果你慢慢走（低速），前面的人有時間反應並側身讓你過，人群的「密度」不會改變。這就是不可壓縮。
    * 如果你用時速 $100$ $\text{km}$ 往前狂衝（接近聲速），前面的人根本來不及躲開，你就會把一堆人撞成一團，導致那邊的人群「密度」瞬間飆高。這就是可壓縮的激波（Shockwave）。
  


## 效果

### 效果 NWP不用處理聲波 delta t 可以很大
同學，你的記憶力非常敏銳！你腦海中浮現的這個「把密度簡化成只跟特定變數有關」的印象是完全正確的。

不過在大氣動力學中，這裡面其實藏著**兩個不同等級的近似法**。嚴格來說，你印象中的 $\rho(x,y,z,t) \to \rho(z)$，其實是比 Boussinesq 更進階的**「滯彈近似 (Anelastic Approximation)」**！

### 第一層次：Boussinesq 近似 (淺層流體)

**核心假設：$\rho(x,y,z,t) \to \rho_0$ (完全是個常數！)**

* **物理情境：** 想像你在煮一鍋淺淺的味噌湯，或是研究貼近地表幾百公尺的邊界層。這裡的流體很淺，所以「背景的大氣壓力」幾乎沒有把底層的空氣壓得特別密。
* **數學操作：** Boussinesq 近似非常霸道，它宣稱流體的密度到處都一樣，完全是一個常數 $\rho_0$（例如 $1.2 \text{ kg/m}^3$）。
* **唯一的例外（浮力項）：** 如果密度到處都一樣，熱空氣怎麼上升？Boussinesq 規定，**只有在計算垂直方向的「重力/浮力」時，才允許密度有一點點微小的擾動 $\rho'$**。在其他所有的動量項、連續方程式裡，$\rho$ 通通被當作死硬的常數 $\rho_0$ 提出去消掉。
    * 連續方程式會退化成最簡單的不可壓縮流：$\nabla \cdot \vec{V} = 0$（也就是 $\frac{\partial u}{\partial x} + \frac{\partial v}{\partial y} + \frac{\partial w}{\partial z} = 0$）。

### 第二層次：滯彈近似 Anelastic Approximation (深層對流/颱風)

**核心假設：$\rho(x,y,z,t) \to \bar{\rho}(z)$ (只隨高度變化的背景值)**

* **物理情境：** 這次我們看的是「颱風」！颱風從海平面一路長到 15 公里高的對流層頂。海平面的空氣很稠密，但 15 公里高的空氣非常稀薄。如果這時候還用 Boussinesq 假設全體密度是常數 $\rho_0$，誤差會大到算不下去。
* **數學操作：** 這就是你記憶中的那個近似！我們把真實的密度拆成兩部分：
    $$\rho(x,y,z,t) = \bar{\rho}(z) + \rho'(x,y,z,t)$$
    * $\bar{\rho}(z)$ 是**背景平均密度**，它只隨高度 $z$ 遞減，不隨時間和水平位置改變。這就像是畫布的底色。
    * $\rho'$ 是因為冷暖造成的**微小擾動**。
* **關鍵捨棄：** 在動量方程式與連續方程式中，我們把主要的慣性質量替換成 $\bar{\rho}(z)$。所以，密度確實從一個四維變數 $\rho(x,y,z,t)$，被我們強行降維成了只隨高度變化的 $\bar{\rho}(z)$！

---

### 為什麼我們要不約而同地丟掉 $\frac{\partial \rho}{\partial t}$？(消音工程)

你有沒有發現，不管是 Boussinesq 還是滯彈近似，我們都絕對不允許背景密度隨「時間 $t$」改變？
也就是我們強迫 $\frac{\partial \rho}{\partial t} = 0$。

* **這步在做什麼？** 這是在把大氣系統裡的**「聲波 (Sound Waves)」**給消音！
* **物理理由：** 聲波的本質就是空氣密度的快速壓縮與膨脹（也就是 $\frac{\partial \rho}{\partial t}$ 不為零）。聲波的速度是 340 m/s，而颱風的風速頂多 50 m/s。如果我們的方程式裡保有 $\frac{\partial \rho}{\partial t}$，在跑電腦數值模式時，為了解析跑得超快的聲波，我們的 Time Step（時間步長）必須切得極度細碎（可能每 0.1 秒就要算一次），這會讓電腦算到當機。
* **結果：** 把 $\frac{\partial \rho}{\partial t}$ 丟掉後，我們等於宣告空氣對聲波來說是「不可壓縮」的，成功把聲波這個討厭的高頻雜訊從方程式裡過濾掉了！

### 回到講義的連續方程式

[cite_start]你可以看看講義第八頁 [cite: 52] 的連續方程式：
$$\frac{\partial\rho ru}{r\partial r}+\frac{\partial\rho w}{\partial z}=0$$

你看這裡面的 $\rho$，它沒有被提出來消掉（因為它不是 Boussinesq 的常數），它被關在偏微分 $\frac{\partial}{\partial z}$ 裡面！這證明了這份講義的颱風模型，使用的是你記憶中的**滯彈近似 (Anelastic)**，允許密度隨高度 $z$ 改變（$\rho = \bar{\rho}(z)$），所以 $\rho$ 遇到對 $z$ 的微分時不能隨便當常數提出來。
