## 一、 核心概念與基本假設 (Core Concept & Assumptions)

* **理論概念：** CISK 是一個描述大尺度環境（近中心邊界層輻合）與小尺度積雲（近中心中層潛熱釋放）互相合作的「尺度交互作用」正回饋機制
* **關鍵數學假設（積雲參數化）：** 寫下加熱率與垂直速度的關係式 $Q \propto w_E \cdot q$。註明這個理論認為能量來源取決於「埃克曼抽吸速度」

### 一、 垂直方向的 3 層劃分 (動力與熱力機制的垂直分層)

在垂直方向上，流體靜力平衡控制了大局，每一層負責不同的質量傳輸與熱力轉換：

1.  **邊界層 (地表至 $\sim 850\text{ hPa}$ 或 $1\text{ km}$)：** 動量耗散與水氣收集層。這裡是唯一摩擦力 ($\nabla \times \vec{\tau}_0$) 不可忽略的地方，負責破壞地轉平衡產生徑向流入 (Radial Inflow)。
2.  **中下對流層 ($\sim 750\text{ hPa}$ 至 $500\text{ hPa}$)：** 潛熱釋放與渦度增強層。積雲在此層發展，潛熱開始大量釋放，且因為質量輻合產生氣旋式渦度 (Spin-up)。
3.  **上對流層 ($\sim 200\text{ hPa}$)：** 質量輻散與高壓排氣層。受下方潛熱加熱導致氣層厚度增加，形成局部高空高壓，負責將質量向外圍拋出 (Radial Outflow)。

### 二、 水平方向的 2 區劃分 (積雲尺度的空間分離)

在水平方向上，CISK 理論區分了「有積雲對流發生的區域」與「無對流發生的環境區域」：

1.  **內部對流區 ($r < 100\text{ km}$)：** 垂直速度大於零 ($w > 0$) 的區域，相當於颱風的眼牆與核心雲帶。這裡是非絕熱加熱 ($Q > 0$) 的主戰場。
2.  **外部環境/沉降區 ($r > 100\text{ km}$)：** 垂直速度小於或等於零 ($w \le 0$) 的區域。提供水氣來源，並透過微弱的下沉運動來補償核心區的質量消耗。

當這 6 個區域建立後，CISK 理論的數學封閉性（如何自己運轉）就非常清晰了：

1. **(外部邊界層)** 輸送高濕度空氣至 **(內部邊界層)**。
2. **(內部邊界層)** 透過摩擦輻合 $w_E = \frac{1}{\rho f} \hat{k} \cdot (\nabla \times \vec{\tau}_0)$ 將水氣往上送入 **(內部中層)**。
3. **(內部中層)** 發生參數化加熱 $Q \propto w_E$，加熱整個氣柱。
4. 氣柱加熱導致 **(內部上層)** 產生高壓與強輻散，進一步使得地表氣壓下降（依據流體靜力方程式 $\frac{\partial p}{\partial z} = -\rho g$ 積分）。
5. 地表低壓加深，使得 **(外部邊界層)** 的流入更強，這又加大了 **(內部邊界層)** 的 $w_E$，完成正回饋。
6. **(外部上層)** 與 **(外部中層)** 則負責閉合這個二次環流，確保質量守恆。
