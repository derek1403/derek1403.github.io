# **Conditional Instability of the Second Kind (CISK, 第二類條件不穩定性)**



## 一、 核心概念與基本假設 (Core Concept & Assumptions)

* **理論概念：** CISK 是一個描述大尺度環境（近中心邊界層輻合）與小尺度積雲（近中心中層潛熱釋放）互相合作的「尺度交互作用」正回饋機制
  * 適用於熱帶氣旋的 **生成初期（Genesis）** 、**弱颱風（熱帶性低氣壓至輕度颱風階段）**
* **關鍵數學假設（積雲參數化）：** 假設整層氣柱的非絕熱加熱率 $Q$ 直接正比於邊界層頂的埃克曼抽吸速度 $w_E$ 與邊界層比濕 $q$ 的乘積。 $Q \propto w_E \cdot q$
  * **物理意義：** 此參數化關係式界定了系統的能量來源，明確指出積雲對流的發展、潛熱釋放量，完全由大尺度系統透過邊界層摩擦所產生的 **Ekman pumping** 來決定與控制

## 二、 空間架構：6 個功能區域 (Spatial Framework)

* **水平劃分：** 
  * 內部對流區 $(r < 100\text{ km})$
  * 外部環境區 $(r > 100\text{ km})$
* **垂直劃分：** 
  * 邊界層 $(\text{地表} \sim 850\text{ hPa})$
  * 中層 $(750\text{ hPa} \sim 200\text{ hPa})$
  * 上層 $(200\text{ hPa} \sim 100\text{ hPa})$


| | 內部對流區 ($r < 100\text{ km}$) | 外部環境區 ($r > 100\text{ km}$) |
| :--- | :--- | :--- |
| **上層**  | **【動力排氣閥】強輻散區**<br> 大量潛熱加熱導致此區等壓面抬升，產生極強的徑向向外氣壓梯度力，氣流在此強烈輻散流出，維持內部邊界層的低壓不被填塞 | **【高空質量接收與下沉起點】**<br>接收來自上層內部的流出質量，流出的空氣在此逐漸輻合，並在科氏力的作用下轉向形成高空反氣旋環流，隨後開始下沉。 |
| **中層**  | **【熱力引擎核心】強烈上升與加熱區**<br>邊界層送上來的水氣在此大量凝結，釋放巨大的潛熱 ($Q \propto w_E \cdot q$)，強烈的上升運動 **拉伸了流體渦管** ，導致此層的相對渦度急遽增加 | **【質量補償與抑制區】微弱下沉區**<br>為了滿足連續方程式的質量守恆，此區產生廣泛且微弱的絕熱下沉運動，下沉增溫使得外圍環境大氣層結趨於穩定，抑制了外圍的對流，確保能量集中於內部對流區。 |
| **邊界層** | **【抽吸泵浦】Ekman pumping**<br>強烈的氣旋式風切與地表摩擦產生最強的邊界層輻合，這裡的垂直速度 ($w_E$) 達到最大值，將匯聚的潮濕空氣像強行向上推入內部中層 | **【水氣與角動量收集區】向心流入區**<br>氣流在向中心輻合的過程中，一方面不斷從溫暖的洋面吸收水氣與潛熱（增加相當位溫 $\theta_e$），另一方面將外圍的絕對角動量向中心輸送。 |

## 三、 動力與熱力耦合：正回饋機制 (Positive Feedback Mechanism)

1. `內部邊界層` 存在初始低壓，使得有梯度風從 `外部邊界層` 吹向 `內部邊界層`
2. `外部邊界層` 輸送高濕度空氣至 `內部邊界層`
3. `內部邊界層` 透過摩擦輻合 $w_E = \frac{1}{\rho f} \hat{k} \cdot (\nabla \times \vec{\tau}_0)$ 將水氣往上送入 `內部中層`
4. `內部中層` 發生參數化加熱 $Q \propto w_E$，加熱整個氣柱。
5. 氣柱加熱導致 `內部上層` 產生高壓與強輻散，進一步使得 `內部邊界層` 氣壓下降（依據流體靜力方程式 $\frac{\partial p}{\partial z} = -\rho g$ 積分）。
6. `內部邊界層` 低壓加深，使得 `外部邊界層` 的流入更強，這又加大了 `內部邊界層` 的 $w_E$，完成正回饋。

*  `外部上層` 與 `外部中層` 則負責閉合這個二次環流，確保質量守恆。




## 四、 理論侷限與現代觀點比較 (Limitations & Modern Perspectives)

科學是不斷修正的過程，記錄理論的缺陷與後續發展，能展現出宏觀的學術視野。

* **短波災難 (Short-wave Catastrophe)：** 點出純線性 CISK 理論的數學缺陷（波數 $k \to \infty$ 時，成長率 $\sigma \to \infty$），這也是為什麼後來 Ooyama 必須人為規定 $r = 100\text{ km}$ 的內區。
* **與 WISHE 理論的對比：**
    * **摩擦力的角色：** CISK 認為是產生輻合的推手；WISHE 認為是消耗動能的阻力。
    * **能量來源的根本：** CISK 依賴垂直抽吸收集現有水氣；WISHE 強調強風掠過海面引發的「表面潛熱通量急遽蒸發」。
    * **熱力學差異：** 點出卡諾循環模型（Carnot Cycle）中，邊界層向心流入的等溫吸熱假設。
