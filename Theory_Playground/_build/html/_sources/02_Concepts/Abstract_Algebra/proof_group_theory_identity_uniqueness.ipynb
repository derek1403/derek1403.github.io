# Proof: The Identity Element in a Group is Unique

## 數學定理小推導 —— 單位元素是唯一的嗎？

群有其中一個定義:

>「There is an element $e \in G$ such that $a = a * e = e * a$ for each $a \in G$」

這句話聽起來像是有可能存在好幾個不同的 $e$，但實際上一個群的單位元素永遠是獨一無二的。

證明目標： 證明群 $G$ 中的單位元素唯一。


推導：



1. 反證法起手： 假設這個群有兩個不同的單位元素，我們分別叫它們 $e_1$ 和 $e_2$。

2. 利用 $e_1$ 的定義： 既然 $e_1$ 是單位元素，那麼任何元素碰到它都不變。我們讓 $e_2$ 去碰到它，可以得到：

$$e_1 * e_2 = e_2$$

3. 利用 $e_2$ 的定義： 同理，既然 $e_2$ 也是單位元素，任何元素碰到它也不變。我們讓 $e_1$ 去碰到它，可以得到：

$$e_1 * e_2 = e_1$$

4. 結論： 觀察上面兩個式子，等號左邊是一模一樣的，所以等號右邊必然相等，也就是推導出：

$$e_1 = e_2$$

這就證明了，無論你一開始假設有幾個單位元素，它們最終都會被證明是同一個東西。
