# Proof of Group Fundamental Theory


## Proof: The Identity Element in a Group is Unique

證明目標： 證明群 $G$ 中的單位元素唯一。

**推導：**

1. 反證法起手： 假設這個群有兩個不同的單位元素，我們分別叫它們 $e_1$ 和 $e_2$。

2. 利用 $e_1$ 的定義： 既然 $e_1$ 是單位元素，那麼任何元素碰到它都不變。我們讓 $e_2$ 去碰到它，可以得到：

$$e_1 * e_2 = e_2$$

3. 利用 $e_2$ 的定義： 同理，既然 $e_2$ 也是單位元素，任何元素碰到它也不變。我們讓 $e_1$ 去碰到它，可以得到：

$$e_1 * e_2 = e_1$$

4. 結論： 觀察上面兩個式子，等號左邊是一模一樣的，所以等號右邊必然相等，也就是推導出：

$$e_1 = e_2$$

這就證明了，無論你一開始假設有幾個單位元素，它們最終都會被證明是同一個東西。

## Proof: The Inverse Element in a Group is Unique

證明目標： 證明元素 $a \in G$ 中的反元素唯一。

**推導：**

假設 $b$ 和 $c$ 是 $a$ 的反元素，那麼:

$$\begin{gather*}
b &=& b * e \\
&=& b * (a * c)  \\
&=& (b * a) * c \\
&=& e * c \\
&=& c
\end{gather*}$$

* $a$ 的反元素記做 $a^{-1}$ 


## Proof : Inverse of an Inverse (反元素的反元素)

**證明目標：** 證明對於所有 $a \in G$，其反元素的反元素 $(a^{-1})^{-1} = a$。

**推導：**
根據群的定義，任何元素 $x$ 若要成為 $a^{-1}$ 的反元素，必須滿足與它相乘等於單位元素 $e$，即 $x * a^{-1} = a^{-1} * x = e$。

$$\begin{gather*}
a^{-1}(a^{-1})^{-1} &=&  e \\
a a^{-1}(a^{-1})^{-1}&=& a   \\
e(a^{-1})^{-1}&=& a \\
(a^{-1})^{-1}&=& a\\
\end{gather*}$$


觀察上式，這正好說明了 $a$ 本身就完美符合「作為 $a^{-1}$ 反元素」的條件。
由於群中的反元素必定是唯一的，因此 $a^{-1}$ 的反元素只能是 $a$，即：


$$(a^{-1})^{-1} = a$$

## Proof 4: Inverse of a Product (乘積的反元素)

**證明目標：** 證明對於所有 $a, b \in G$，相乘後的反元素 $(a * b)^{-1} = b^{-1} * a^{-1}$。

**推導：**

$$\begin{gather*}
(a * b)*(a * b)^{-1} &=&  e \\
a^{-1}*(a * b)*(a * b)^{-1} &=&  a^{-1}*e \\
b*(a * b)^{-1}&=& a^{-1}   \\
b^{-1}*b*(a * b)^{-1}&=& b^{-1}*a^{-1}   \\
(a * b)^{-1}&=& b^{-1}*a^{-1}  \\
\end{gather*}$$


要證明 $b^{-1} * a^{-1}$ 確實是 $(a * b)$ 的反元素，我們必須證明兩者相乘的結果為單位元素 $e$。
利用群的結合律，我們將它們相乘：

$$\begin{gather*}
&&(a * b) * (b^{-1} * a^{-1})  \\
&=&a *e*a^{-1}  \\
&=& e \\
\end{gather*}$$

同理，我們也檢查反方向的相乘：

$$\begin{gather*}
&&(b^{-1} * a^{-1})*(a * b)   \\
&=&b *e*b^{-1}  \\
&=& e \\
\end{gather*}$$

既然正向與反向相乘的結果皆為 $e$，根據反元素的唯一性，可知 $(a * b)$ 的反元素必定是 $b^{-1} * a^{-1}$。

## Proof 5a: Unique Solutions for Equations (方程式的唯一解)

**證明目標：** 證明對於所有 $a, b \in G$，方程式 $a * x = b$ （與 $y * a = b$）在群 $G$ 中有唯一解。

**推導：**
我們以 $a * x = b$ 為例，這可以分為兩步來嚴格證明：

**1. 證明解存在 (Existence)：**
因為 $a \in G$，所以必定存在 $a^{-1}$。我們構造一個解 $x = a^{-1} * b$ （由於封閉性，此解必定存在於 $G$ 中）。
將其代入方程式左側檢驗：


$$a * (a^{-1} * b) = (a * a^{-1}) * b = e * b = b$$


等式成立，證明至少存在一個解。

**2. 證明解唯一 (Uniqueness)：**
假設有兩個解 $x_1$ 與 $x_2$ 都滿足條件，即 $a * x_1 = b$ 且 $a * x_2 = b$。
這表示 $a * x_1 = a * x_2$。我們在等式兩邊的「左側」同時乘上 $a^{-1}$：


$$a^{-1} * (a * x_1) = a^{-1} * (a * x_2)$$


利用結合律展開：


$$(a^{-1} * a) * x_1 = (a^{-1} * a) * x_2$$

$$e * x_1 = e * x_2$$

$$x_1 = x_2$$


因為任意兩個解最終必定相等，故證明該方程式的解是唯一的。（方程式 $y * a = b$ 的證明邏輯完全相同，只需在右側同乘 $a^{-1}$ 即可）。

## Proof 5b: Cancellation Laws (消去律)

**證明目標：** 證明如果 $a * b = a * c$，則推導出 $b = c$；如果 $a * b = c * b$，則推導出 $a = c$。

**推導：**
我們以左消去律（Left cancellation） $a * b = a * c$ 為例。
由於 $a \in G$，其反元素 $a^{-1}$ 必定存在。我們在等式兩側的「左邊」同乘 $a^{-1}$：


$$a^{-1} * (a * b) = a^{-1} * (a * c)$$


利用結合律將括號重新組合：


$$(a^{-1} * a) * b = (a^{-1} * a) * c$$

$$e * b = e * c$$

$$b = c$$

同理，對於右消去律（Right cancellation） $a * b = c * b$，我們只需在等式的「右邊」同時乘上 $b^{-1}$，便能以完全相同的步驟推導出 $a = c$。
