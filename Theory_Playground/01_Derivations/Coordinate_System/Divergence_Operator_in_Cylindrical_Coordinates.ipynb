# Divergence Operator in Cylindrical Coordinates (圓柱座標中的散度算子)



## 證明目標:

對於任意向量場 $\mathbf{A} = A_r\hat{r} + A_\lambda\hat{\lambda} + A_z\hat{z}$，證明其散度為：


$$
\nabla \cdot \mathbf{A} = \frac{1}{r}\frac{\partial[r A_r]}{\partial r} + \frac{1}{r}\frac{\partial A_\lambda}{\partial \lambda} + \frac{\partial A_z}{\partial z}
$$



## 假設與已知 (Assumptions & Preliminaries)

* **【已知 1】 圓柱座標的 $\nabla$ (Nabla) 算子定義：**
  $$\nabla = \hat{r}\frac{\partial}{\partial r} + \hat{\lambda}\frac{1}{r}\frac{\partial}{\partial \lambda} + \hat{z}\frac{\partial}{\partial z}$$

* **【已知 2】 [圓柱座標單位向量的直角座標轉換](https://derek1403.github.io/Theory_Playground/_build/html/01_Derivations/Coordinate_System/cylindrical_coordinate_transformation_momentum_equation.html#step-2-local-basis-objects)：**
  
  直角座標的單位向量 $(\hat{I}, \hat{J}, \hat{K})$ 是絕對固定的常數，但圓柱座標的單位向量 $(\hat{r}, \hat{\lambda}, \hat{z})$ 會隨著角度 $\lambda$ 變化：

  * $\hat{r} = \cos\lambda \hat{I} + \sin\lambda \hat{J}$
  * $\hat{\lambda} = -\sin\lambda \hat{I} + \cos\lambda \hat{J}$
  * $\hat{z} = \hat{K}$

* **【已知 3】 單位向量的偏微分：**
  

  $$\begin{bmatrix}
  \frac{\partial \hat{r}}{\partial r}       & \frac{\partial \hat{r}}{\partial \lambda}       & \frac{\partial \hat{r}}{\partial z} \\
  \frac{\partial \hat{\lambda}}{\partial r} & \frac{\partial \hat{\lambda}}{\partial \lambda} & \frac{\partial \hat{\lambda}}{\partial z} \\
  \frac{\partial \hat{z}}{\partial r}       & \frac{\partial \hat{z}}{\partial \lambda}       & \frac{\partial \hat{z}}{\partial z}
  \end{bmatrix} = \begin{bmatrix}
  0 & \mathbf{\hat{\lambda}} & 0 \\
  0 & \mathbf{-\hat{r}}      & 0 \\
  0 & 0                      & 0 
  \end{bmatrix}$$


  利用【已知 2】，我們對 $\hat{r}$ 和 $\hat{\lambda}$ 取偏微分：
  * $\frac{\partial \hat{r}}{\partial \lambda} = -\sin\lambda \hat{i} + \cos\lambda \hat{j} = \mathbf{\hat{\lambda}}$  *(這代表徑向向量轉動時，變化量的方向切好是切線方向)*
  * $\frac{\partial \hat{\lambda}}{\partial \lambda} = -\cos\lambda \hat{i} - \sin\lambda \hat{j} = \mathbf{-\hat{r}}$ *(這代表切線向量轉動時，變化量的方向是指向圓心的)*

* **【已知 4】 基底是正交的：**
  
  $$\begin{bmatrix}
  \hat{r} \cdot \hat{r}      & \hat{\lambda} \cdot \hat{r}        & \hat{z} \cdot \hat{r}       \\
  \hat{r} \cdot \hat{\lambda} &  \hat{\lambda} \cdot \hat{\lambda}& \hat{z} \cdot \hat{\lambda} \\
  \hat{r} \cdot \hat{z}       & \hat{\lambda} \cdot \hat{r}       & \hat{z} \cdot \hat{z}
  \end{bmatrix} = \begin{bmatrix}
  1 & 0 & 0 \\
  0 & 1 & 0 \\
  0 & 0 & 1 
  \end{bmatrix}$$



## 證明:

我們將 $\nabla$ 算子與向量 $\mathbf{A}$ 做內積 (Dot Product)。這是一條主線，我們只需要耐心把分配律展開：

$$\begin{gather*}
\nabla \cdot \mathbf{A} &=&& \left( \hat{r}\frac{\partial}{\partial r} + \hat{\lambda}\frac{1}{r}\frac{\partial}{\partial \lambda} + \hat{z}\frac{\partial}{\partial z} \right) \cdot \left( A_r\hat{r} + A_\lambda\hat{\lambda} + A_z\hat{z} \right) \\
&=&&  \hat{r} \cdot \frac{\partial}{\partial r} \left( A_r\hat{r} + A_\lambda\hat{\lambda} + A_z\hat{z} \right) \\
&&+& \frac{1}{r} \hat{\lambda} \cdot \frac{\partial}{\partial \lambda} \left( A_r\hat{r} + A_\lambda\hat{\lambda} + A_z\hat{z} \right)\\
&&+&  \hat{z} \cdot \frac{\partial}{\partial z} \left( A_r\hat{r} + A_\lambda\hat{\lambda} + A_z\hat{z} \right)\\
&=&&  \hat{r} \cdot \left[ \left(\frac{\partial A_r}{\partial r}\hat{r} + A_r\frac{\partial \hat{r}}{\partial r}\right) + \left(\frac{\partial A_\lambda}{\partial r}\hat{\lambda} + A_\lambda\frac{\partial \hat{\lambda}}{\partial r}\right) + \left(\frac{\partial A_z}{\partial r}\hat{z} + A_z\frac{\partial \hat{z}}{\partial r}\right) \right] \\
&&+& \frac{1}{r} \hat{\lambda} \cdot \left[ \left(\frac{\partial A_r}{\partial \lambda}\hat{r} + A_r\frac{\partial \hat{r}}{\partial \lambda}\right) + \left(\frac{\partial A_\lambda}{\partial \lambda}\hat{\lambda} + A_\lambda\frac{\partial \hat{\lambda}}{\partial \lambda}\right) + \left(\frac{\partial A_z}{\partial \lambda}\hat{z} + A_z\frac{\partial \hat{z}}{\partial \lambda}\right) \right]\\
&&+&  \hat{z} \cdot \left[ \left(\frac{\partial A_r}{\partial z}\hat{r} + A_r\frac{\partial \hat{r}}{\partial z}\right) + \left(\frac{\partial A_\lambda}{\partial z}\hat{\lambda} + A_\lambda\frac{\partial \hat{\lambda}}{\partial z}\right) + \left(\frac{\partial A_z}{\partial z}\hat{z} + A_z\frac{\partial \hat{z}}{\partial z}\right) \right] \\
&\overset{\text{已知 3}}{=}&&  \hat{r} \cdot \left[ \left(\frac{\partial A_r}{\partial r}\hat{r} + A_r0\right) + \left(\frac{\partial A_\lambda}{\partial r}\hat{\lambda} + A_\lambda 0\right) + \left(\frac{\partial A_z}{\partial r}\hat{z} + A_z 0\right) \right] \\
&&+& \frac{1}{r} \hat{\lambda} \cdot \left[ \left(\frac{\partial A_r}{\partial \lambda}\hat{r} + A_r\hat{\lambda}\right) + \left(\frac{\partial A_\lambda}{\partial \lambda}\hat{\lambda} + A_\lambda (-\hat{r}) \right) + \left(\frac{\partial A_z}{\partial \lambda}\hat{z} + A_z 0\right) \right]\\
&&+&  \hat{z} \cdot \left[ \left(\frac{\partial A_r}{\partial z}\hat{r} + A_r 0\right) + \left(\frac{\partial A_\lambda}{\partial z}\hat{\lambda} + A_\lambda 0\right) + \left(\frac{\partial A_z}{\partial z}\hat{z} + A_z 0\right) \right] \\
&=&&  \hat{r} \cdot \left[ \frac{\partial A_r}{\partial r}\hat{r}  + \frac{\partial A_\lambda}{\partial r}\hat{\lambda} + \frac{\partial A_z}{\partial r}\hat{z} \right] \\
&&+& \frac{1}{r} \hat{\lambda} \cdot \left[ \left(\frac{\partial A_r}{\partial \lambda}\hat{r} + A_r\hat{\lambda}\right) + \left(\frac{\partial A_\lambda}{\partial \lambda}\hat{\lambda} + A_\lambda (-\hat{r}) \right) + \left(\frac{\partial A_z}{\partial \lambda}\hat{z} \right) \right]\\
&&+&  \hat{z} \cdot \left[ \frac{\partial A_r}{\partial z}\hat{r}  + \frac{\partial A_\lambda}{\partial z}\hat{\lambda}   + \frac{\partial A_z}{\partial z}\hat{z}  \right] \\
&=&&   \frac{\partial A_r}{\partial r}\hat{r} \cdot \hat{r}  + \frac{\partial A_\lambda}{\partial r}\hat{r} \cdot \hat{\lambda} +\frac{\partial A_z}{\partial r}\hat{r} \cdot \hat{z}  \\
&&+& \frac{1}{r}  \left[ \left(\frac{\partial A_r}{\partial \lambda}\hat{\lambda} \cdot \hat{r} + A_r\hat{\lambda} \cdot \hat{\lambda} \right)+ \left( \frac{\partial A_\lambda}{\partial \lambda}\hat{\lambda} \cdot \hat{\lambda} + A_\lambda \hat{\lambda} \cdot (-\hat{r})  \right)+ \frac{\partial A_z}{\partial \lambda}\hat{\lambda} \cdot \hat{z}  \right]\\
&&+&  \frac{\partial A_r}{\partial z}\hat{z} \cdot\hat{r}  + \frac{\partial A_\lambda}{\partial z}\hat{z} \cdot\hat{\lambda}  + \frac{\partial A_z}{\partial z}\hat{z} \cdot\hat{z}   \\
&\overset{\text{已知 4}}{=}&&   \frac{\partial A_r}{\partial r}  + \frac{\partial A_\lambda}{\partial r}0 +\frac{\partial A_z}{\partial r}0  \\
&&+& \frac{1}{r}  \left[ \left(\frac{\partial A_r}{\partial \lambda}0 + A_r \right)+ \left( \frac{\partial A_\lambda}{\partial \lambda} + A_\lambda 0  \right)+ \frac{\partial A_z}{\partial \lambda}0  \right]\\
&&+&  \frac{\partial A_r}{\partial z}0  + \frac{\partial A_\lambda}{\partial z}0  + \frac{\partial A_z}{\partial z}   \\
&=&&   \frac{\partial A_r}{\partial r}   +  \frac{1}{r}  \left[  A_r +  \frac{\partial A_\lambda}{\partial \lambda}  \right]+ \frac{\partial A_z}{\partial z}   \\
&=&&   \frac{\partial A_r}{\partial r}   +  \frac{1}{r}   A_r +  \frac{1}{r}\frac{\partial A_\lambda}{\partial \lambda}  + \frac{\partial A_z}{\partial z}   \\
&=&&   \left(\frac{\partial A_r}{\partial r}   +  \frac{1}{r}   A_r \right)+  \frac{1}{r}\frac{\partial A_\lambda}{\partial \lambda}  + \frac{\partial A_z}{\partial z}   \\
&=&&    \frac{1}{r} \left(r\frac{\partial A_r}{\partial r}   +    A_r \right)+  \frac{1}{r}\frac{\partial A_\lambda}{\partial \lambda}  + \frac{\partial A_z}{\partial z}   \\
&=&&   \frac{1}{r} \frac{\partial[r A_r]}{\partial r}+  \frac{1}{r}\frac{\partial A_\lambda}{\partial \lambda}  + \frac{\partial A_z}{\partial z}   \\
\end{gather*}$$
