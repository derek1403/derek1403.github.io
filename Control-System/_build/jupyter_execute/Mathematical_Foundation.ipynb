## Laplace Transform and Inverse Laplace Transform

1. Laplace Transform 用處 :

    把時間域(t-domain)的微分方程轉換成複數域(s-domain)的代數方程，微分變成$s$、積分變成$\frac{1}{s}$的乘法解方程就方便許多。

    在控制系統中描述`線性時非系統(Linear Time-Invariant System ,LTI System)`的`傳遞函數(Transfer function ,G)`並用來分析系統的穩定性、響應特性。

  $$
  G(s) = \frac{Y(s)}{U(s)} \tag{eq.2.1}
  $$


  ```{note}
    線性非時變系統簡單來說就是可以將一個狀態分解成幾個組成的線性疊加，例如說可以分解成由外力引起的響應+由初始條件引起的響應=總體的響應。

    我覺得從控制的角度Laplace Transform是Fourier transform的升級版，相當於拓展了卷積定理，存在$e^{+t}$傅立葉會發散而拉普拉斯可以收斂。
    
    Laplace Transform拿到傳遞函數並在虛軸上求頻率響應:奈奎斯特圖(Nyquist plot)、伯德圖(Bold plot)。

    Laplace Transform在數學上可以看作是矩量生成函數(moment generating function, MGF)的一種形式。
  ```


2. Laplace Transform :

  * Condition :

    要求$f(t)$不能成長的跟$e^{t^{2}}$一樣快，也就是要滿足 :

  $$
  |f(t)| \leq M e^{\sigma t}, \quad t \geq 0 ,\quad \sigma \in \mathbb{R}
  \\
  \int_{0}^{\infty} \big| f(t) e^{-\sigma t} \big| \, dt < \infty ,\quad \sigma \in \mathbb{R}
  $$

  * Definition :

    將時域函數$f(t)$映射到以複數平面上的各個頻率為基底的空間，並組合出$F(s)$。

  $$
  \begin{align*}
  F(s)  &\overset{def}{=} \int_{0}^{\infty}  f(t) e^{-s t}  \, dt ,\quad s \in \mathbb{C}   \\
   &\overset{def}{=} \mathcal{L}[f(t)] \\
  \end{align*}
  $$

3. Inverse Laplace Transform :

  * Definition :

    將函數$F(s)$沿積分路徑是一條平行於虛軸的直線積分收斂區域得出$f(t)$。

  $$
  \begin{align*}
  f(t)  &\overset{def}{=} \frac{1}{2\pi j} \int_{\gamma - j\infty}^{\gamma + j\infty}  F(s) e^{st} \, ds \\
  &\overset{def}{=} \mathcal{L}^{-1}[F(s)](t)
  \end{align*}
  $$

  ```{note}
    j是虛數單位，這堂課比較常出現j代表虛數單位。
    這個定義建議是看看就好 :3
  ```

4. 常用轉換表與規則

$$
\begin{array}{|c|c|}
\hline
\text{時域函數 } f(t) & \text{頻域函數 } F(s) = \mathcal{L}\{f(t)\} \\
\hline
1 & \dfrac{1}{s}, \quad \Re(s) > 0 \\
\hline
t & \dfrac{1}{s^2}, \quad \Re(s) > 0 \\
\hline
t^n & \dfrac{n!}{s^{n+1}}, \quad \Re(s) > 0 \\
\hline
e^{at} & \dfrac{1}{s-a}, \quad \Re(s) > a \\
\hline
\sin(bt) & \dfrac{b}{s^2 + b^2}, \quad \Re(s) > 0 \\
\hline
\cos(bt) & \dfrac{s}{s^2 + b^2}, \quad \Re(s) > 0 \\
\hline
e^{at}\sin(bt) & \dfrac{b}{(s-a)^2 + b^2}, \quad \Re(s) > a \\
\hline
e^{at}\cos(bt) & \dfrac{s-a}{(s-a)^2 + b^2}, \quad \Re(s) > a \\
\hline
\delta(t) & 1 \\
\hline
u_{s}(t) \,  & \dfrac{1}{s}, \quad \Re(s) > 0 \\
\hline
\end{array}
$$

* 線性 (Linearity)

$$
\mathcal{L}\{k f(t)\} = kF(s) \\
\mathcal{L}\{f_1(t) + f_2(t)\} = F_1(s) + F_2(s) \\
$$

* 導數 (Differentiation)

$$
\mathcal{L}\left\{\frac{df}{dt}\right\} = sF(s) - f(0^+) \\
\mathcal{L}\left\{\frac{d^n f}{dt^n}\right\} = s^n F(s) - s^{n-1}f(0^+) - \cdots - f^{(n-1)}(0^+) \\
$$

* 積分 (Integration)

$$
\mathcal{L}\left\{\int_{0}^{t} f(\tau)\, d\tau \right\} = \frac{1}{s} F(s) \\
$$

* 時移 (Time Shifting)

$$
\mathcal{L}\{ f(t-T) u(t-T) \} = e^{-sT} F(s) \\
$$

* **頻移 (Frequency Shifting)**

$$
\mathcal{L}\{ e^{-at} f(t) \} = F(s+a) \\
$$

* 卷積 (Convolution)

$$
\mathcal{L}\left\{\int_{0}^{t} f_1(\tau) f_2(t-\tau)\, d\tau \right\} = F_1(s) \cdot F_2(s) \\
$$

* 初值定理 (Initial Value Theorem

$$
\lim_{t \to 0^+} f(t) = \lim_{s \to \infty} sF(s) \\
$$

* 終值定理 (Final Value Theorem)
$$
\lim_{t \to \infty} f(t) = \lim_{s \to 0} sF(s) \quad (\text{系統穩定時})
$$

5. 例題

  $$
  y'' + 5y' + 6y = u(t),
  \quad
  \left\{
  \begin{array}{l}
  y(0) = -2, \\
  y'(0) = 5, \\
  u(t) = 2u_s(t)
  \end{array}
  \right.
  $$
     
  * 解法1(ODE) :

    計算$y_h(t)$和$y_p(t)$並合併，帶入初值。

  $$
  \lambda^2+5\lambda+6=0 \quad\Rightarrow\quad (\lambda+3)(\lambda+2)=0 \\
  y_h(t)=C_1 e^{-3t}+C_2 e^{-2t}
  $$


  $$
  6y_p = 2 \quad\Rightarrow\quad y_p=\frac{1}{3}
  $$

  $$
  \underline{y(t) = -\frac{1}{3}e^{-3t} - 2e^{-2t} + \frac{1}{3},\quad t\ge 0}
  $$

  * 解法2(Laplace) :

    丟到$\mathcal{L}$再$\mathcal{L}^{-1}$丟回來。

  $$
  \mathcal{L}\{y''+5y'+6y\} = \mathcal{L}\{u(t)\} = \frac{2}{s} \\
  $$

  $$
  \begin{align*}
  \mathcal{L}\{y''\} &= s^2 Y(s) - s y(0) - y'(0) &=& s^2 Y(s) + 2s - 5 \\
  \mathcal{L}\{5y'\} &= 5(sY(s) - y(0)) &=& 5(sY(s)+2) \\
  \mathcal{L}\{6y\} &= 6 Y(s)
  \end{align*}
  $$

  $$
  \begin{align*}
  Y(s) &= \frac{-2s^2 -5s +2}{s(s^2 + 5s +6)}  \\
  &= \frac{\frac{1}{3}}{s} + \frac{-2}{s+2} + \frac{-\frac{1}{3}}{s+3}
  \end{align*}
  $$
  
  $$
  \underline{y(t) = \mathcal{L}^{-1}\{Y(s)\} = -\frac{1}{3} e^{-3t} - 2 e^{-2t} + \frac{1}{3},\quad t\ge 0}
  $$
  




## Bold Plot









````python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal

def plot_bode(G, w_min=1e-2, w_max=1e2, num_points=1000, dB=False):
    """
    繪製任意傳遞函數 G(s) 的 Bode Plot（幅值與相位）。

    參數：
    --------
    G : scipy.signal.TransferFunction
        傳遞函數物件，例如 signal.TransferFunction([1, T], [1])
    w_min, w_max : float
        頻率範圍 (rad/s)
    num_points : int
        頻率取樣點數
    dB : bool
        True 則幅值以 dB 顯示，False 則顯示線性值
    """
    # 頻率軸
    w = np.logspace(np.log10(w_min), np.log10(w_max), num_points)

    # 計算頻率響應
    w, mag, phase = signal.bode(G, w=w)  # mag 已經是 dB，phase 是度

    # 幅值處理
    if not dB:
        mag = 10**(mag/20)  # 轉回線性

    # 畫幅值圖
    plt.figure(figsize=(7.5, 4.8))
    plt.semilogx(w, mag, lw=2, color='C0')
    plt.grid(True, which='both', ls='--', alpha=0.5)
    plt.xlabel(r'Frequency $\omega$ (rad/s)')
    plt.ylabel('|G(jω)|' + (' [dB]' if dB else ''))
    plt.title('Bode Magnitude')
    plt.tight_layout()

    # 畫相位圖
    plt.figure(figsize=(7.5, 4.8))
    plt.semilogx(w, phase, lw=2, color='C2')
    plt.grid(True, which='both', ls='--', alpha=0.5)
    plt.xlabel(r'Frequency $\omega$ (rad/s)')
    plt.ylabel('Phase (deg)')
    plt.title('Bode Phase')
    plt.tight_layout()

    plt.show()

# ===== 使用範例 =====
# G(s) = 1 + sT, 假設 T = 1
T = 1.0
num = [1, T]   # 分子係數
den = [1]      # 分母係數
G = signal.TransferFunction(num, den)

plot_bode(G, dB=False)  # 線性幅值
plot_bode(G, dB=True)   # dB 幅值
````
